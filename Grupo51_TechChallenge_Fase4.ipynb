### 1.3 Exibição de logs - CustomLogging
 - Habilita ou desabilita o logs para fins de monitoria ou debug


```python
# 1.3 - Exibição de Gravação de Logs - Classe CustomLogging

LOG_SHOW_CONSOLE = False  # Ativar exibição de log na tela
LOG_ENABLE_FILE = False   # Ativar gravação de log em arquivo
LOG_DEBUG = False         # Ativar exibição de log de processamento do frame

!pip install pytz

import logging
from datetime import datetime
import pytz
from typing import List
import os


LOG_NAME_FILE = "Grupo51TechChallengeFase4.log"

class CustomLogger:
    def __init__(self,
                 name: str = "CustomLogger",
                 log_to_file: bool = False,
                 log_file: str = "app.log",
                 show_console: bool = True,
                 log_debug: bool = False):

        self.logger = logging.getLogger(name)
        self.logger.setLevel(logging.INFO)
        self.logs: List[str] = []
        self.enabled = show_console | log_debug
        self.log_to_file = log_to_file
        self.log_file = log_file
        self.show_console = show_console
        self.log_debug = log_debug

         # Configurar timezone Brasil
        self.tz_br = pytz.timezone('America/Sao_Paulo')

        # Classe para converter tempo para BR
        class BRFormatter(logging.Formatter):
            def converter(self, timestamp):
                dt = datetime.fromtimestamp(timestamp)
                return dt.astimezone(pytz.timezone('America/Sao_Paulo'))

            def formatTime(self, record, datefmt=None):
                dt = self.converter(record.created)
                if datefmt:
                    return dt.strftime(datefmt)
                return dt.strftime('%Y-%m-%d %H:%M:%S')

        # Limpa handlers existentes
        self.logger.handlers.clear()

        # Configurar formatter
        self.formatter =  BRFormatter('[%(levelname)s][%(asctime)s]: %(message)s', datefmt='%d-%m-%Y %H:%M:%S')

        # Handler para console se ativado
        if self.show_console:
            self._setup_console_handler()

        # Handler para arquivo se ativado
        if self.log_to_file:
            self._setup_file_handler()

    def _setup_console_handler(self):
        """Configura o handler para console"""
        self.console_handler = logging.StreamHandler()
        self.console_handler.setLevel(logging.INFO)
        self.console_handler.setFormatter(self.formatter)
        self.logger.addHandler(self.console_handler)

    def _setup_file_handler(self):
        """Configura o handler para arquivo"""
        self.file_handler = logging.FileHandler(self.log_file)
        self.file_handler.setLevel(logging.INFO)
        self.file_handler.setFormatter(self.formatter)
        self.logger.addHandler(self.file_handler)

    def enable_console(self):
        """Ativa a exibição de logs no console"""
        self.show_console = True
        self._setup_console_handler()

    def disable_console(self):
        """Desativa a exibição de logs no console"""
        self.show_console = False
        for handler in self.logger.handlers[:]:
            if isinstance(handler, logging.StreamHandler) and not isinstance(handler, logging.FileHandler):
                self.logger.removeHandler(handler)

    def enable_file_logging(self, log_file: str = None):
        """Ativa o logging em arquivo"""
        if log_file:
            self.log_file = log_file
        self.log_to_file = True
        self._setup_file_handler()

    def disable_file_logging(self):
        """Desativa o logging em arquivo"""
        self.log_to_file = False
        for handler in self.logger.handlers[:]:
            if isinstance(handler, logging.FileHandler):
                self.logger.removeHandler(handler)

    def enable(self):
        """Ativa o registro de logs"""
        self.enabled = True

    def disable(self):
        """Desativa o registro de logs"""
        self.enabled = False

    def debug(self, message: str):
        """Registra uma mensagem de debug"""

        if self.log_debug:
           print(f"[DEBUG]", message)

    def log(self, message: str, level: str = "INFO"):
        """Registra uma mensagem de log"""
        if not self.enabled:
            return

        timestamp = datetime.now().strftime("%d-%m-%Y %H:%M:%S")
        log_message = f"[{level}] {timestamp} {message}"
        self.logs.append(log_message)

        if level == "INFO":
            self.logger.info(message)
        elif level == "WARNING":
            self.logger.warning(message)
        elif level == "ERROR":
            self.logger.error(message)
        elif level == "FRBUG":
            self.logger.debug(message)

    def show_logs(self):
        """Exibe todos os logs armazenados"""
        if not self.logs:
            print("Nenhum log registrado.")
            return

        print("\n=== LOGS ===")
        for log in self.logs:
            print(log)
        print("===========")

    def show_file_logs(self):
        """Exibe logs do arquivo"""
        if not self.log_to_file or not os.path.exists(self.log_file):
            print("Arquivo de log não encontrado ou logging em arquivo desativado.")
            return

        print(f"\n=== LOGS DO ARQUIVO: {self.log_file} ===")
        with open(self.log_file, 'r') as f:
            print(f.read())
        print("===========")

    def clear_logs(self):
        """Limpa todos os logs armazenados"""
        self.logs.clear()

    def clear_file_logs(self):
        """Limpa o arquivo de logs"""
        if self.log_to_file and os.path.exists(self.log_file):
            open(self.log_file, 'w').close()

# Criar logger com todas as opções
logger = CustomLogger(name="",
                      log_to_file=LOG_ENABLE_FILE,
                      log_file=LOG_NAME_FILE,
                      show_console=LOG_SHOW_CONSOLE,
                      log_debug = LOG_DEBUG)

logger.log("Log ativado")
```

### 1.3.1 Classe LoggingManager


```python
# 1.3.1 Classe LoggingManager

import logging
import sys
import traceback
from datetime import datetime
from pathlib import Path
from typing import Optional
from IPython.display import display, HTML

class LoggerManager:
    """
    Gerenciador de logs otimizado para Google Colab
    """

    COLORS = {
        'DEBUG': '#4B8BBE',    # Azul Python
        'INFO': '#45B39D',     # Verde
        'WARNING': '#F4D03F',  # Amarelo
        'ERROR': '#E74C3C',    # Vermelho
        'CRITICAL': '#C0392B', # Vermelho escuro
    }

    def __init__(self,
                 logger_name: str,
                 log_level: int = logging.DEBUG,
                 flag_log_console: bool = True,
                 flag_log_file: bool = False,
                 log_dir: Optional[str] = None,
                 log_filename: Optional[str] = None):
        """
        Inicializa o gerenciador de logs

        Args:
            logger_name: Nome do logger
            log_level: Nível de logging (DEBUG, INFO, etc)
            flag_log_console: Se True, exibe logs no Colab
            flag_log_file: Se True, grava logs em arquivo
            log_dir: Diretório para arquivos de log
            log_filename: Nome do arquivo de log (opcional)
        """
        self.logger = logging.getLogger(logger_name)
        self.logger.setLevel(log_level)
        self.logger.handlers = []

        self.log_format = '%(asctime)s - %(levelname)s - %(message)s'

        if flag_log_console:
            self._setup_colab_logging()

        if flag_log_file:
            self._setup_file_logging(log_dir, log_filename)

    def _setup_colab_logging(self):
        """Configura logging para o Colab"""
        colab_handler = self._ColabOutputHandler()
        colab_handler.setFormatter(self._ColabFormatter(self.log_format))
        self.logger.addHandler(colab_handler)

    def _setup_file_logging(self, log_dir: Optional[str], log_filename: Optional[str]):
        """Configura logging para arquivo"""
        if log_dir:
            Path(log_dir).mkdir(parents=True, exist_ok=True)

        if not log_filename:
            log_filename = f'log_{datetime.now().strftime("%Y%m%d_%H%M%S")}.log'

        file_path = str(Path(log_dir if log_dir else '.') / log_filename)
        file_handler = logging.FileHandler(file_path)
        file_handler.setFormatter(logging.Formatter(self.log_format))
        self.logger.addHandler(file_handler)

    class _ColabFormatter(logging.Formatter):
        """Formatador personalizado para o Colab com HTML"""
        def format(self, record):
            log_message = super().format(record)
            color = LoggerManager.COLORS.get(record.levelname, 'black')
            font_weight = 'bold' if record.levelname in ['ERROR', 'CRITICAL'] else 'normal'

            return f"""
                <div style="
                    color: {color};
                    font-family: monospace;
                    font-weight: {font_weight};
                    margin: 2px 0;
                    padding: 2px;
                    border-left: 3px solid {color};
                    padding-left: 10px;
                ">
                    {log_message}
                </div>
            """

    class _ColabOutputHandler(logging.Handler):
        """Handler personalizado para output no Colab"""
        def emit(self, record):
            log_entry = self.format(record)
            display(HTML(log_entry))

    def debug(self, message: str):
        """Log de debug"""
        self.logger.debug(message)

    def info(self, message: str):
        """Log de informação"""
        self.logger.info(message)

    def warning(self, message: str):
        """Log de aviso"""
        self.logger.warning(message)

    def error(self, message: str):
        """Log de erro"""
        self.logger.error(message)
        if sys.exc_info()[0]:
            self.logger.error(f"Traceback: {traceback.format_exc()}")

    def critical(self, message: str):
        """Log crítico"""
        self.logger.critical(message)
        if sys.exc_info()[0]:
            self.logger.critical(f"Traceback: {traceback.format_exc()}")
```

### 1.8 - Gerenciamento de Arquivos - Classe FileManager

```python
# 1.9 - Classe FileManager

from datetime import datetime
import pytz

class FileManager:
   """
   Classe responsável por gerenciar operações com arquivos e fornecer informações detalhadas
   """
   def __init__(self):
       self.video_extensions = ['.mp4', '.avi', '.mov', '.mkv']
       self.supported_codecs = ['MP4V', 'XVID', 'avc1']

       logger.log('Chamando FileManager...')

   def mount_drive(self):
        # Monta o drive, se o mesmo ainda não estiver montado
        drive_path = '/content/drive'
        if not os.path.exists(f"{drive_path}/MyDrive"):
            print("Montando o Google Drive...")
            drive.mount(drive_path)
        else:
            logger.log("Google Drive já esta montado.")

   def get_file_details(self, file_path):
       """
       Obtém detalhes do arquivo
       Args:
           file_path: Caminho do arquivo
       Returns:
           dict: Dicionário com detalhes do arquivo
       """
       try:
           if not os.path.exists(file_path):
               return {
                   'exists': False,
                   'error': f"Arquivo não encontrado: {file_path}"
               }

           file_stats = os.stat(file_path)
           file_extension = os.path.splitext(file_path)[1].lower()

           details = {
               'exists': True,
               'path': file_path,
               'name': os.path.basename(file_path),
               'extension': file_extension,
               'size_bytes': file_stats.st_size,
               'size_mb': round(file_stats.st_size / (1024 * 1024), 2),
               'created': datetime.fromtimestamp(file_stats.st_ctime).strftime('%Y-%m-%d %H:%M:%S'),
               'modified': datetime.fromtimestamp(file_stats.st_mtime).strftime('%Y-%m-%d %H:%M:%S'),
               'is_video': file_extension in self.video_extensions
           }

           # Se for vídeo, adiciona detalhes específicos
           if details['is_video']:
               video_details = self._get_video_details(file_path)
               details.update(video_details)

           return details

       except Exception as e:
           return {
               'exists': False,
               'error': f"Erro ao obter detalhes: {str(e)}"
           }

   def _get_video_details(self, video_path):
       """
       Obtém detalhes específicos de arquivos de vídeo
       Args:
           video_path: Caminho do vídeo
       Returns:
           dict: Detalhes do vídeo
       """
       try:
           cap = cv2.VideoCapture(video_path)
           if not cap.isOpened():
               return {'error': 'Não foi possível abrir o vídeo'}

           details = {
               'frame_count': int(cap.get(cv2.CAP_PROP_FRAME_COUNT)),
               'fps': int(cap.get(cv2.CAP_PROP_FPS)),
               'width': int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
               'height': int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)),
               'duration_seconds': int(cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS))
           }

           cap.release()
           return details

       except Exception as e:
           return {'error': f"Erro ao obter detalhes do vídeo: {str(e)}"}

   def read_and_show_details(self, file_path):
       """
       Lê e exibe os detalhes do arquivo
       Args:
           file_path: Caminho do arquivo
       """
       try:
           if not os.path.exists(file_path):
               print(f"Arquivo não encontrado: {file_path}")
               return

           # Obtém informações do arquivo
           file_stats = os.stat(file_path)
           file_extension = os.path.splitext(file_path)[1].lower()

           # Exibe informações básicas do arquivo
           print("\nInformações do Arquivo:")
           print("-" * 40)
           print(f"Nome: {os.path.basename(file_path)}")
           print(f"Caminho: {file_path}")
           print(f"Extensão: {file_extension}")
           print(f"Tamanho: {round(file_stats.st_size / (1024 * 1024), 2)} MB")
           print(f"Criado em: {datetime.fromtimestamp(file_stats.st_ctime).strftime('%Y-%m-%d %H:%M:%S')}")
           print(f"Modificado em: {datetime.fromtimestamp(file_stats.st_mtime).strftime('%Y-%m-%d %H:%M:%S')}")

           # Se for vídeo, exibe informações adicionais
           if file_extension in self.video_extensions:
               cap = cv2.VideoCapture(file_path)
               if cap.isOpened():
                   print("\nDetalhes do Vídeo:")
                   print("-" * 40)
                   print(f"Resolução: {int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))}x{int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))}")
                   print(f"FPS: {int(cap.get(cv2.CAP_PROP_FPS))}")
                   total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
                   duration = total_frames / cap.get(cv2.CAP_PROP_FPS)
                   print(f"Total de Frames: {total_frames}")
                   print(f"Duração: {int(duration)} segundos ({round(duration/60, 2)} minutos)")
                   cap.release()

           print("-" * 40)

       except Exception as e:
           print(f"Erro ao ler arquivo: {str(e)}")

   def generate_filename(self,base_name, extensao):
      """Generates a filename with the specified pattern and current timestamp in Brazilian timezone.
      Args:
        base_name: The base name for the file.
      Returns:
        A string representing the generated filename.
      """
      now = datetime.now(pytz.timezone('America/Sao_Paulo'))
      formatted_time = now.strftime("%d%m%Y_%H%M%S")

      return f"{base_name}_{formatted_time}.{extensao}"

   def display_files(self, input_video_path, processed_video_path, report_path):
        """Displays the paths of the input video, processed video, and report file."""

        print("Arquivo video de entrada:", input_video_path)
        print("Arquivo de video analisado:", processed_video_path)
        print("Arquivo de relatório de saida:", report_path)

   def validate_video(self, video_path):
       """
       Valida se o arquivo de vídeo é válido e pode ser processado
       Args:
           video_path: Caminho do vídeo
       Returns:
           bool: True se válido, False caso contrário
       """
       try:
           if not os.path.exists(video_path):
               print(f"Arquivo não encontrado: {video_path}")
               return False

           if os.path.splitext(video_path)[1].lower() not in self.video_extensions:
               print(f"Formato de arquivo não suportado. Use: {self.video_extensions}")
               return False

           cap = cv2.VideoCapture(video_path)
           if not cap.isOpened():
               print("Não foi possível abrir o vídeo")
               return False

           # Verifica parâmetros básicos
           fps = int(cap.get(cv2.CAP_PROP_FPS))
           width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
           height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
           frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

           cap.release()

           if fps <= 0 or width <= 0 or height <= 0 or frame_count <= 0:
               print("Vídeo com parâmetros inválidos")
               return False

           return True

       except Exception as e:
           print(f"Erro ao validar vídeo: {str(e)}")
           return False

   def compare_videos(self, original_path, processed_path):
       """
       Compara dois arquivos de vídeo e mostra as diferenças
       Args:
           original_path: Caminho do vídeo original
           processed_path: Caminho do vídeo processado
       """
       try:
           print("\nComparação de Vídeos:")
           print("-" * 40)

           # Obtém detalhes dos vídeos
           original_details = self.get_file_details(original_path)
           processed_details = self.get_file_details(processed_path)

           # Compara tamanhos
           size_diff = processed_details['size_mb'] - original_details['size_mb']
           size_ratio = (processed_details['size_mb'] / original_details['size_mb']) * 100

           print("Tamanho dos arquivos:")
           print(f"- Original: {original_details['size_mb']:.2f} MB")
           print(f"- Processado: {processed_details['size_mb']:.2f} MB")
           print(f"- Diferença: {size_diff:+.2f} MB ({size_ratio:.1f}%)")

           # Compara parâmetros de vídeo
           print("\nParâmetros de vídeo:")
           print(f"- Resolução Original: {original_details['width']}x{original_details['height']}")
           print(f"- Resolução Processado: {processed_details['width']}x{processed_details['height']}")
           print(f"- FPS Original: {original_details['fps']}")
           print(f"- FPS Processado: {processed_details['fps']}")
           print(f"- Duração Original: {original_details['duration_seconds']} segundos")
           print(f"- Duração Processado: {processed_details['duration_seconds']} segundos")

           print("-" * 40)

       except Exception as e:
           print(f"Erro ao comparar vídeos: {str(e)}")
```

```python
import cv2
import mediapipe as mp
import numpy as np
import math

class BodyMovement:
    def __init__(self):
        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(
            min_detection_confidence=0.5,
            min_tracking_confidence=0.5,
            model_complexity=1
        )

        # Dicionário para armazenar o histórico de pontos
        self.previous_landmarks = {}

        # Dicionário para armazenar os contadores de movimentos
        self.movement_counters = {
            "braco_levantado": 0,
            "mao_gesticulando": 0,
            "caminhando": 0,
            "em_pe": 0,
            "sentado": 0,
            "cumprimento_maos": 0,
            "esfregando_maos": 0,
            "gesticulando_boca": 0,
            "maos_para_cima": 0
        }

        # Limiares para detecção de movimentos
        self.thresholds = {
            "movimento_geral": 0.02,
            "braco_levantado": 0.3,
            "mao_gesticulando": 0.05,
            "caminhando": 0.1,
            "sentado": 0.4,
            "em_pe": 0.7,
            "cumprimento_maos": 0.1,
            "esfregando_maos": 0.15,
            "gesticulando_boca": 0.02,
            "maos_para_cima": 0.6
        }

    def calculate_distance(self, point1, point2):
        """Calcula a distância euclidiana entre dois pontos"""
        return math.sqrt((point1.x - point2.x)**2 + (point1.y - point2.y)**2)

    def calculate_angle(self, point1, point2, point3):
        """Calcula o ângulo entre três pontos"""
        a = np.array([point1.x, point1.y])
        b = np.array([point2.x, point2.y])
        c = np.array([point3.x, point3.y])

        ba = a - b
        bc = c - b

        cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
        angle = np.arccos(cosine_angle)

        return np.degrees(angle)

    # body-1
    def draw_body_movements(self, frame, body_movements):
        try:
            print ('[BodyMovement][draw_body_movements] Desenhando movimento corporal')
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = self.pose.process(frame_rgb)

            if not results or not results.pose_landmarks:
                return frame

            landmarks = results.pose_landmarks.landmark
            h, w, _ = frame.shape

            def landmark_to_pixel(landmark_index):
                try:
                    landmark = landmarks[landmark_index]
                    return int(landmark.x * w), int(landmark.y * h)
                except:
                    return None

            # Desenha contadores no canto superior esquerdo
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.6
            thickness = 2
            text_x = 10
            text_y = 30
            line_height = 25

            if results.pose_landmarks:
                # Braço Levantado - Linha amarela
                if body_movements.get("braco_levantado", 0) > 0:
                    for side in ["RIGHT", "LEFT"]:
                        shoulder = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_SHOULDER").value)
                        wrist = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_WRIST").value)
                        if shoulder and wrist:
                            cv2.line(frame, shoulder, wrist, (0, 255, 255), 2)
                            # Adiciona texto na linha
                            mid_point = ((shoulder[0] + wrist[0]) // 2, (shoulder[1] + wrist[1]) // 2)
                            cv2.putText(frame, "Braco Levantado", mid_point, font, 0.5, (0, 255, 255), 2)

                # Gesticulando - Círculos vermelhos
                if body_movements.get("mao_gesticulando", 0) > 0:
                    wrist_right = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_WRIST.value)
                    wrist_left = landmark_to_pixel(self.mp_pose.PoseLandmark.LEFT_WRIST.value)
                    if wrist_right and wrist_left:
                        cv2.circle(frame, wrist_right, 10, (0, 0, 255), -1)
                        cv2.circle(frame, wrist_left, 10, (0, 0, 255), -1)
                        mid_point = ((wrist_right[0] + wrist_left[0]) // 2, (wrist_right[1] + wrist_left[1]) // 2)
                        cv2.putText(frame, "Gesticulando", mid_point, font, 0.5, (0, 0, 255), 2)

                # Caminhando - Linhas roxas
                if body_movements.get("caminhando", 0) > 0:
                    for side in ["RIGHT", "LEFT"]:
                        hip = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_HIP").value)
                        ankle = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_ANKLE").value)
                        if hip and ankle:
                            cv2.line(frame, hip, ankle, (255, 0, 255), 2)
                            mid_point = ((hip[0] + ankle[0]) // 2, (hip[1] + ankle[1]) // 2)
                            cv2.putText(frame, "Caminhando", mid_point, font, 0.5, (255, 0, 255), 2)

                # Em Pé - Linha verde
                if body_movements.get("em_pe", 0) > 0:
                    nose = landmark_to_pixel(self.mp_pose.PoseLandmark.NOSE.value)
                    right_ankle = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_ANKLE.value)
                    if nose and right_ankle:
                        cv2.line(frame, nose, right_ankle, (0, 255, 0), 2)
                        mid_point = ((nose[0] + right_ankle[0]) // 2, (nose[1] + right_ankle[1]) // 2)
                        cv2.putText(frame, "Em Pe", mid_point, font, 0.5, (0, 255, 0), 2)

                # Sentado - Retângulo azul
                if body_movements.get("sentado", 0) > 0:
                    hip = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_HIP.value)
                    knee = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_KNEE.value)
                    if hip and knee:
                        cv2.rectangle(frame, hip, knee, (255, 0, 0), 2)
                        mid_point = ((hip[0] + knee[0]) // 2, (hip[1] + knee[1]) // 2)
                        cv2.putText(frame, "Sentado", mid_point, font, 0.5, (255, 0, 0), 2)

                # Cumprimento - Círculo ciano
                if body_movements.get("cumprimento_maos", 0) > 0:
                    wrist_right = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_WRIST.value)
                    wrist_left = landmark_to_pixel(self.mp_pose.PoseLandmark.LEFT_WRIST.value)
                    if wrist_right and wrist_left:
                        mid_point = ((wrist_right[0] + wrist_left[0]) // 2, (wrist_right[1] + wrist_left[1]) // 2)
                        cv2.circle(frame, mid_point, 15, (255, 255, 0), -1)
                        cv2.putText(frame, "Cumprimento", (mid_point[0], mid_point[1] - 20), font, 0.5, (255, 255, 0), 2)

                # Esfregando Mãos - Linhas verdes entre as mãos
                if body_movements.get("esfregando_maos", 0) > 0:
                    wrist_right = landmark_to_pixel(self.mp_pose.PoseLandmark.RIGHT_WRIST.value)
                    wrist_left = landmark_to_pixel(self.mp_pose.PoseLandmark.LEFT_WRIST.value)
                    if wrist_right and wrist_left:
                        cv2.line(frame, wrist_right, wrist_left, (0, 255, 0), 2)
                        mid_point = ((wrist_right[0] + wrist_left[0]) // 2, (wrist_right[1] + wrist_left[1]) // 2)
                        cv2.putText(frame, "Esfregando Maos", mid_point, font, 0.5, (0, 255, 0), 2)

                # Gesticulando com a Boca - Retângulo rosa
                if body_movements.get("gesticulando_boca", 0) > 0:
                    mouth_right = landmark_to_pixel(self.mp_pose.PoseLandmark.MOUTH_RIGHT.value)
                    mouth_left = landmark_to_pixel(self.mp_pose.PoseLandmark.MOUTH_LEFT.value)
                    if mouth_right and mouth_left:
                        cv2.rectangle(
                            frame, (mouth_right[0] - 10, mouth_right[1] - 10), (mouth_left[0] + 10, mouth_left[1] + 10), (255, 128, 255), 2
                        )
                        text_pos = (mouth_right[0], mouth_right[1] - 20)
                        cv2.putText(frame, "Gesticulando Boca", text_pos, font, 0.5, (255, 128, 255), 2)

                # Mãos para Cima - Linhas amarelas
                if body_movements.get("maos_para_cima", 0) > 0:
                    for side in ["LEFT", "RIGHT"]:
                        shoulder = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_SHOULDER").value)
                        wrist = landmark_to_pixel(getattr(self.mp_pose.PoseLandmark, f"{side}_WRIST").value)
                        if shoulder and wrist:
                            cv2.line(frame, shoulder, wrist, (255, 255, 0), 3)
                            mid_point = ((shoulder[0] + wrist[0]) // 2, (shoulder[1] + wrist[1]) // 2)
                            cv2.putText(frame, "Maos para Cima", mid_point, font, 0.5, (255, 255, 0), 2)

            return frame

        except Exception as e:
            print(f"BodyMovement: Erro ao desenhar movimentos corporais: {str(e)}")
            return frame
```

```python
# 2.2 - Tratamento de movimentos corporais - Classe BodyMovement
import cv2
import mediapipe as mp
import numpy as np

class BodyMovementDetector:
    def __init__(self):
        """Inicializa o detector de movimentos corporais"""
        # Inicializa MediaPipe para detecção de pose
        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(
            static_image_mode=False,
            model_complexity=1,
            min_detection_confidence=0.5,
            min_tracking_confidence=0.5
        )

        logger.log('Chamando BodyMovementDetector...')

        # Definir limiares para detecção de movimentos
        self.POSE_THRESHOLD = 0.3
        self.MOVEMENT_THRESHOLD = 0.15
        self.ANGLE_THRESHOLD = 0.2

        # Estado anterior para comparação
        self.previous_state = None

    # body 1
    def detect_movements(self, frame):
        """
        Detecta movimentos corporais no frame
        Args:
            frame: Frame do vídeo
        Returns:
            dict: Dicionário com os movimentos detectados
        """
        try:
            landmarks = []

            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = self.pose.process(frame_rgb)

            movements = {
                'braco_levantado': 0,
                'mao_gesticulando': 0,
                'caminhando': 0,
                'sentado': 0,
                'em_pe': 0,
                'deitado': 0,
                'aperto_maos':0
            }

            if results.pose_landmarks:
                landmarks = results.pose_landmarks.landmark

                # Detecta braço levantado
                movements['braco_levantado'] = self._detect_raised_arm(landmarks)

                # Detecta gestos com as mãos
                movements['mao_gesticulando'] = self._detect_hand_gesture(landmarks)

                # Detecta caminhada
                movements['caminhando'] = self._detect_walking(landmarks)

                # Detecta posição sentada
                movements['sentado'] = self._detect_sitting(landmarks)

                # Detecta posição em pé
                movements['em_pe'] = self._detect_standing(landmarks)

                # Detecta posição deitada
                movements['deitado'] = self._detect_lying(landmarks)

            self.previous_state = movements
            return movements

        except Exception as e:
            print(f"PoseMovementAnalyzer: Erro na detecção de movimentos 2: {str(e)}")
            return {k: 0 for k in ['braco_levantado', 'mao_gesticulando', 'caminhando',
                                 'sentado', 'em_pe', 'deitado']}

    def _detect_raised_arm(self, landmarks):
        """Detecta braço levantado"""
        try:
            shoulder_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y
            wrist_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_WRIST.value].y
            return 1 if (shoulder_y - wrist_y) > self.POSE_THRESHOLD else 0
        except:
            return 0

    def _detect_hand_gesture(self, landmarks):
        """Detecta gestos com as mãos"""
        try:
            wrist_x = landmarks[self.mp_pose.PoseLandmark.RIGHT_WRIST.value].x
            elbow_x = landmarks[self.mp_pose.PoseLandmark.RIGHT_ELBOW.value].x
            return 1 if abs(wrist_x - elbow_x) > self.MOVEMENT_THRESHOLD else 0
        except:
            return 0

    def _detect_walking(self, landmarks):
        """Detecta caminhada"""
        try:
            left_ankle_y = landmarks[self.mp_pose.PoseLandmark.LEFT_ANKLE.value].y
            right_ankle_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_ANKLE.value].y
            return 1 if abs(left_ankle_y - right_ankle_y) > self.MOVEMENT_THRESHOLD else 0
        except:
            return 0

    def _detect_sitting(self, landmarks):
        """Detecta posição sentada"""
        try:
            hip_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_HIP.value].y
            knee_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_KNEE.value].y
            ankle_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

            knees_bent = (knee_y - ankle_y) < self.POSE_THRESHOLD
            hip_position = (hip_y - knee_y) < self.POSE_THRESHOLD

            return 1 if knees_bent and hip_position else 0
        except:
            return 0

    def _detect_standing(self, landmarks):
        """Detecta posição em pé"""
        try:
            nose_y = landmarks[self.mp_pose.PoseLandmark.NOSE.value].y
            ankle_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_ANKLE.value].y
            return 1 if (ankle_y - nose_y) > self.POSE_THRESHOLD else 0
        except:
            return 0

    def _detect_lying(self, landmarks):
        """Detecta posição deitada"""
        try:
            left_shoulder_y = landmarks[self.mp_pose.PoseLandmark.LEFT_SHOULDER.value].y
            right_shoulder_y = landmarks[self.mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y
            return 1 if abs(left_shoulder_y - right_shoulder_y) < self.ANGLE_THRESHOLD else 0
        except:
            return 0

    # BodyMovementDetector
    def draw_movements(self, frame, movements):
        """
        Desenha indicadores visuais dos movimentos detectados
        Args:
            frame: Frame do vídeo
            movements: Dicionário com os movimentos detectados
        Returns:
            frame: Frame com as marcações
        """
        try:

            # Importa os módulos necessários
            mp_drawing = mp.solutions.drawing_utils
            mp_pose = mp.solutions.pose

            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = self.pose.process(frame_rgb)

            if not results.pose_landmarks:
                return frame

            h, w, _ = frame.shape
            landmarks = results.pose_landmarks.landmark

            # Função auxiliar para converter landmark em pixel
            def landmark_to_pixel(landmark_index):
                try:
                    landmark = landmarks[landmark_index]
                    return (int(landmark.x * w), int(landmark.y * h))
                except:
                    return None

            # Desenha indicadores para cada movimento detectado
            y_position = 30
            for movement, value in movements.items():
                if value > 0:
                    color = {
                        'braco_levantado': (0, 255, 255),  # Amarelo
                        'mao_gesticulando': (0, 0, 255),   # Vermelho
                        'caminhando': (255, 0, 255),       # Magenta
                        'sentado': (255, 0, 0),            # Azul
                        'em_pe': (0, 255, 0),              # Verde
                        'deitado': (0, 165, 255),          # Laranja
                        'aperto_maos': (0, 165, 255)       # Laranja

                    }.get(movement, (255, 255, 255))       # Branco (default)

                    cv2.putText(frame, f"{movement}: {value}",
                              (10, y_position), cv2.FONT_HERSHEY_SIMPLEX,
                              0.7, color, 2)
                    y_position += 30

            return frame

        except Exception as e:
            print(f"BodyMovementDetector: Erro ao desenhar movimentos: {str(e)}")
            return frame

```

```python
# 2.4 - Análise atividades anômalas - Classe AnomalyMovementDetector

import numpy as np

class AnomalyMovementDetector:
    def __init__(self):

        self._logger = CustomLogger(
            show_console=  LOG_SHOW_CONSOLE,
            log_debug= LOG_DEBUG
           )

        self._logger.debug("[AnomalyMovementDetector] Iniciando...")

        self.np = np

        # Inicializa MediaPipe
        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(
            min_detection_confidence=0.5,
            min_tracking_confidence=0.5
             )

        self.previous_landmarks = None

        self.velocity_thresholds = {
            'low': 1.0,
            'medium': 2.5,
            'high': float('inf')
        }
        self.acceleration_thresholds = {
            'low': 1.5,
            'medium': 3.0,
            'high': float('inf')
        }

        # Inicializa dicionário de anomalias
        self.default_anomalies = {
            # Status atual
            'movimento_brusco': False,
            'tremor': False,
            'queda': False,
            'desequilibrio': False,
            'colisao': False,
            'desmaio': False,

            # Contadores totais (histórico completo)
            'total_movimento_brusco': 0,
            'total_tremor': 0,
            'total_queda': 0,
            'total_desequilibrio': 0,
            'total_colisao': 0,
            'total_desmaio': 0,

            # Contadores da sessão atual
            'sessao_movimento_brusco': 0,
            'sessao_tremor': 0,
            'sessao_queda': 0,
            'sessao_desequilibrio': 0,
            'sessao_colisao': 0,
            'sessao_desmaio': 0,

            # Métricas gerais
            'intensidade': 0.0,
            'confianca': 0.0,
            'movimento_anomalo': False,
            'total_anomalias': 0,
            'total_anomalias_sessao': 0,
            'tempo_inicio_sessao': None,
            'duracao_sessao': 0
        }

        # Inicializa contadores de anomalias com os mesmos tipos
        self.anomaly_counts = self.default_anomalies.copy()

        self.movement_history = []
        self.anomaly_threshold = {
            'speed': 2.0,  # Threshold para velocidade anormal
            'pattern': 0.7, # Threshold para padrões de movimento
            'sudden_change': 0.5  # Threshold para mudanças bruscas
        }

        self.window_size = 10  # Tamanho da janela para análise temporal

        self.velocity_threshold = 0.01    # Limiar muito baixo para garantir detecção
        self.tremor_threshold = 0.01     # Limiar muito baixo para garantir detecção
        self.fall_threshold = 0.01       # Limiar muito baixo para garantir detecção

        # Histórico para suavização
        self.position_history = []
        self.history_size = 5

    def _get_default_movements(self):
        """
        Retorna um dicionário com valores padrão para todos os tipos
        de movimentos anômalos monitorados
        """
        return {
            # Movimentos básicos
            'velocidade': 0.0,
            'movimento_vertical': 0.0,
            'direcao': 0.0,
            'intensidade': 0.0,

        # Movimentos anômalos específicos
            'tremor': False,
            'queda': False,
            'movimento_brusco': False,
            'desequilibrio': False,
            'colisao': False,
            'desmaio': False,

        # Status geral
            'movimento_anomalo': False,

        # Métricas auxiliares
            'confianca': 0.0,
           'tempo_deteccao': 0.0
    }

    # draw_anomalies - melhorada

    def draw_anomalies(self, frame, anomalies):
        """
        Desenha as anomalias detectadas no frame
        """

        self._logger.debug("[AnomalyMovementDetector][draw_anomalies]")

        # Dicionário de posições para cada tipo de anomalia
        positions = {
            "tremor": (10, 30),
            "movimento_brusco": (10, 60),
            "queda": (10, 90),
            "desequilibrio": (10, 120),
            "colisao": (10, 150),
            "desmaio": (10, 180)
        }

        try:
            for anomaly_type, value in anomalies.items():
            # Pula movimento_anomalo pois é um indicador agregado
                if anomaly_type == 'movimento_anomalo':
                    continue

                if value > 0:
                    # Obtém a posição do dicionário ou usa posição padrão se não encontrar
                    position = positions.get(anomaly_type, (10, 210))

                    # Desenha texto
                    cv2.putText(
                       frame,
                        f"ALERTA: {anomaly_type.upper()}",
                        position,
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7,
                        (0, 0, 255),
                      2
                  )

                 # Desenha retângulo de alerta
                    cv2.rectangle(
                      frame,
                      (5, position[1]-25),
                      (300, position[1]+5),
                      (0, 0, 255),
                      2
                 )

            return frame

        except Exception as e:
              print(f"AnomalyMovementDetector: Erro ao desenhar movimentos: '{str(e)}'")
              return frame

    def detect_anomalies(self, landmarks):
        """Detecta diferentes tipos de movimentos anômalos"""

        #print("[AnomalyMovementDetector][detect_anomalies] detectando anomalias")

        anomalies = {
          'tremor': self._detect_tremor(landmarks),
          'queda': self._detect_fail(landmarks),
          'movimento_brusco': self._detect_sudden_movement(landmarks),
          'desequilibrio': self._detect_imbalance(landmarks),
          'colisao': self._detect_collision(landmarks),
          'desmaio': self._detect_faint(landmarks)
          }

        return anomalies


    def _detect_balance_loss(self, movements):
        """Detecta perda de equilíbrio"""
        try:
            equilibrio = movements.get('equilibrio', 0.0)
            return 1 if equilibrio > 0.7 else 0
        except Exception as e:
            print(f"[EnhancedAnomalyDetector] Erro na detecção de desequilíbrio: {str(e)}")
            return 0

    def _detect_speed_anomalies(self, movements):
        """Detecta anomalias de velocidade"""
        try:
            velocidade = movements.get('velocidade', 0.0)
            return 1 if velocidade > 0.5 else 0
        except Exception as e:
            self._logger.debug(f"[AnomalyMovementDetector] Erro na detecção de velocidade: {str(e)}")
            return 0

    def _detect_tremor(self, movements):
        """Detecta tremores"""
        try:
            #self._logger.debug("[AnomalyMovementDetector] detectando tremor")

            intensidade = movements.get('intensidade', 0.0)
            return 1 if intensidade > 0.8 else 0
        except Exception as e:
            self._logger.debug(f"[AnomalyMovementDetector] Erro na detecção de tremor: {str(e)}")
            return 0

    def _detect_fail(self, movements):
        """Detecta quedas"""
        try:
           # print("[EnhancedAnomalyDetector] detectando queda")
            mov_vertical = movements.get('movimento_vertical', 0.0)
            return 1 if mov_vertical > 1.0 else 0
        except Exception as e:
            self._logger.debug(f"[AnomalyMovementDetector][_detect_fail] Erro na detecção de queda: {str(e)}")
            return 0

    def _detect_collision(self, movements):
        """Detecta colisões"""
        try:
          #  self._logger.debug("[AnomalyMovementDetector] detectando colisão")
            mudanca_direcao = movements.get('mudanca_direcao', 0.0)
            return 1 if mudanca_direcao > 0.8 else 0
        except Exception as e:
            self._logger.debug(f"[AnomalyMovementDetector][_detect_collision] Erro na detecção de colisão: {str(e)}")
            return 0

    def _detect_faint(self, movements):
        """Detecta desmaios"""
        try:
            #self._logger.debug("[AnomalyMovementDetector] detectando desmaio")
            movimento_vertical = movements.get('movimento_vertical', 0.0)
            intensidade = movements.get('intensidade', 0.0)
            return 1 if movimento_vertical < 0.1 and intensidade < 0.1 else 0
        except Exception as e:
            self.draw_anomalous_movements_logger.debug(f"[AnomalyMovementDetector][_detect_faint] Erro na detecção de desmaio: {str(e)}")
            return 0


    def draw_anomalous_movements(self, frame, anomalies):
        """Desenha informações sobre movimentos anômalos no frame no lado direito"""

        try:
            if frame is None:
                return frame

            self._logger.debug("[AnomalyMovementDetector] Desenhando movimentos anômalos")

            frame_copy = frame.copy()

            # Posições iniciais para o texto
            pos_x = 875
            pos_y = 30

            # Fonte e configurações
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.6
            font_thickness = 1

          # Cores
            text_color = (0, 255, 0)  # Verde
            warning_color = (0, 0, 255)  # Vermelho

            # Lista completa de informações para exibir
            display_items = [
            ('Tremor', anomalies.get('tremor', False)),
            ('Total Tremor', anomalies.get('total_tremor', 0)),
            ('Queda', anomalies.get('queda', False)),
            ('Total Queda', anomalies.get('total_queda', 0)),
            ('Colisão', anomalies.get('colisao', False)),
            ('Total Colisão', anomalies.get('total_colisao', 0)),
            ('Desmaio', anomalies.get('desmaio', False)),
            ('Total Desmaio', anomalies.get('total_desmaio', 0)),
            ('Desequilíbrio', anomalies.get('desequilibrio', False)),
            ('Total Desequil.', anomalies.get('total_desequilibrio', 0)),
            ('Mov. Anômalo', anomalies.get('movimento_anomalo', False)),
            ('Total Mov. Anômalo', anomalies.get('total_movimento_anomalo', 0)),
            ('Total Anomalias', anomalies.get('total_anomalias', 0)),
            ('Tempo Sessão', anomalies.get('tempo_inicio_sessao', 0))
          ]

        # Desenha cada item
            for label, value in display_items:
                text = f'{label}: {value}'
                color = warning_color if value else text_color

              # Adiciona retângulo de fundo para melhor legibilidade
                (text_width, text_height), _ = cv2.getTextSize(text, font, font_scale, font_thickness)
                cv2.rectangle(frame_copy,
                         (pos_x-5, pos_y-text_height-5),
                         (pos_x+text_width+5, pos_y+5),
                         (0, 0, 0),
                         -1)

                cv2.putText(frame_copy, text, (pos_x, pos_y), font,
                       font_scale, color, font_thickness)
                pos_y += 22  # reduzido ainda mais o espaçamento para caber todas as informações

            return frame_copy

        except Exception as e:
          self._logger.debug(f"[AnomalyMovementDetector] Erro ao desenhar movimentos anômalos: {str(e)}")
          import traceback
          print(f"Traceback: {traceback.format_exc()}")
          return frame
```

### 2.4.1 Classe class EnhancedAnomalyDetector

```python
import numpy as np
from typing import List, Dict, Any, Tuple
from datetime import datetime

import mediapipe as mp

class EnhancedAnomalyDetector:
    def __init__(self,
                 buffer_size: int = 30,
                 speed_threshold: float = 0.5,    # Reduzido de 2.0 para 0.5
                 pattern_threshold: float = 0.3,  # Reduzido de 0.7 para 0.3
                 direction_threshold: float = 0.2, # Reduzido de 0.5 para 0.2
                 window_size: int = 15,           # Reduzido de 30 para 15
                 fall_threshold: float = 0.3      # Reduzido de 0.5 para 0.3
                 ):
        """
        Inicializa o detector de anomalias

        Args:
            speed_threshold: Limiar para detecção de velocidade anormal
            pattern_threshold: Limiar para detecção de padrões anormais
            direction_threshold: Limiar para detecção de mudanças bruscas
            window_size: Tamanho da janela de análise (em frames)
        """

        self._logger = CustomLogger(
                       show_console = LOG_SHOW_CONSOLE,
                       log_debug= LOG_DEBUG)


        self.anomaly_detector = AnomalyMovementDetector()

        self.default_anomalies = {
            'tremor': False,
            'queda': False,
            'colisao': False,
            'delirio': False,
            'desmaio': False,
            'desequilibrio': False,
            'movimento_brusco': False,
            'total_anomalias': 0,
            'total_anomalias_sessao': 0,
            'tempo_inicio_sessao': None,
            'duracao_sessao': 0,

            # Adicionando os novos contadores
            'sessao_movimento_brusco': 0,
            'sessao_tremor': 0,
            'sessao_queda': 0,
            'sessao_desequilibrio': 0,
            'sessao_colisao': 0,
            'sessao_desmaio': 0,
            'total_movimento_brusco': 0,
            'total_tremor': 0,
            'total_queda': 0,
            'total_desequilibrio': 0,
            'total_colisao': 0,
            'total_desmaio': 0
        }

        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5 )

        # Parâmetros de configuração
        self.buffer_size = buffer_size
        self.speed_threshold = speed_threshold
        self.pattern_threshold = pattern_threshold
        self.direction_threshold = direction_threshold
        self.window_size = window_size
        self.fall_threshold = fall_threshold

         # Inicializa previous_landmarks
        self.previous_landmarks = None

        self.acceleration_thresholds = {
            'low': 1.5,
            'medium': 3.0,
            'high': float('inf')
        }

         # Inicializa contadores de anomalias com os mesmos tipos
        self.anomaly_counts = self.default_anomalies.copy()

        # Históricos e contadores
        self.anomaly_history = []
        self.movement_history = []
        self.movement_buffer = []  # Buffer para análise de janela deslizante

         # Contador de anomalias por tipo
        self.anomaly_counters  = {
          'tremor': 0,
          'queda': 0,
          'movimento_brusco': 0,
          'desequilibrio': 0,
          'colisao': 0,
          'desmaio': 0
        }

        self.anomaly_types = {
        'speed': 0,
        'pattern': 0,
        'direction': 0
     }

    def log_anomalies(self, anomalies, frame_id):
        if anomalies.get("movimento_anomalo"):
           detectados = [t for t in ["movimento_brusco", "tremor", "queda", "desequilibrio", "colisao", "desmaio"] if anomalies.get(t)]
           #self._logger.log(f"Frame {frame_id} - Anomalias: {', '.join(detectados)}")


    def _extract_movements(self, frame):
        """
        Extrai métricas de movimento do frame.
        Returns:
            dict: Dicionário com as métricas de movimento
        """
        #self._logger.debug("[EnhancedAnomalyDetector][_extract_movements] Iniciando....")

        try:

            if frame.size > 0:  # verifica se o frame não está vazio
                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            else:
                raise ValueError("Frame vazio recebido")

            results = self.pose.process(frame_rgb)

            if not results.pose_landmarks:

               self._logger.debug("[EnhancedAnomalyDetector][_extract_movements] Nenhum landmark detectado neste frame")
               return {
                  'velocidade': 0.0,
                  'movimento_vertical': 0.0,
                  'equilibrio': 0.0,
                  'mudanca_direcao': 0.0,
                  'intensidade': 0.0
               }

            self._logger.debug("[EnhancedAnomalyDetector][extract_movementos] Landmarks detectados, iniciando cálculos")
            landmarks = results.pose_landmarks.landmark

            # Calcula velocidade
            self._logger.debug("[Enhanced-extract_movementos] Calculando velocidade")
            if not hasattr(self, 'previous_landmarks'):
                self._logger.debug("[Enhanced][_extract_movements] Primeiro frame - inicializando previous_landmarks")
                self.previous_landmarks = landmarks
                velocidade = 0.0
            else:
                velocidade = self._calcular_velocidade(landmarks)
                self._logger.log(f"[Enhanced][_extract_movements] Velocidade calculada: {velocidade}")
                self.previous_landmarks = landmarks

            # Calcula movimento vertical
            #print("[DEBUG] Calculando movimento vertical")
            movimento_vertical = self._calcular_movimento_vertical(landmarks)
          #  print(f"[DEBUG] Movimento vertical calculado: {movimento_vertical}")

            # Calcula equilíbrio
           # print("[DEBUG] Calculando equilíbrio")
            equilibrio = self._calcular_equilibrio(landmarks)
           # print(f"[DEBUG] Equilíbrio calculado: {equilibrio}")

            # Calcula intensidade
           # print("[DEBUG] Calculando intensidade")
            intensidade = self._calcular_intensidade(landmarks)
           # print(f"[DEBUG] Intensidade calculada: {intensidade}")

            movements = {
            'velocidade': velocidade,
            'movimento_vertical': movimento_vertical,
            'equilibrio': equilibrio,
            'mudanca_direcao': 0.0,
            'intensidade': intensidade
               }

            logger.log("[Enhanced][_extract_movements] Movimentos calculados:")
            for key, value in movements.items():
                self._logger.debug(f"[Enhanced][_extract_movements] - {key}: {value}")

            return movements

        except Exception as e:
               self._logger.debug(f"[ERROR][Enhanced][_extract_movements] Erro na extração de movimentos: {str(e)}")
               self._logger.debug(f"[ERROR][Enhanced][_extract_movements] Tipo do erro: {type(e).__name__}")
               import traceback
               self._logger.debug(f"[ERROR][Enhanced][_extract_movements] Traceback completo:\n{traceback.format_exc()}")

               return      {
                 'velocidade': 0.0,
                  'movimento_vertical': 0.0,
                'equilibrio': 0.0,
                'mudanca_direcao': 0.0,
                'intensidade': 0.0
                  }

    def _calcular_velocidade(self, landmarks):
        """
        Calcula a velocidade média do movimento
        """
        #print("[DEBUG] Iniciando cálculo de velocidade")
        try:
            if not hasattr(self, 'previous_landmarks'):
                return 0.0

            velocidade_total = 0
            num_landmarks = len(landmarks)

            for i in range(num_landmarks):
                atual = landmarks[i]
                anterior = self.previous_landmarks[i]

                delta = ((atual.x - anterior.x)**2 +
                       (atual.y - anterior.y)**2)**0.5
                velocidade_total += delta

            velocidade = velocidade_total / num_landmarks
         #   print(f"[DEBUG] Velocidade calculada: {velocidade}")
            return velocidade

        except Exception as e:
            self._logger.log(f"[ERROR] Erro ao calcular velocidade: {str(e)}")
            return 0.0

    def _calcular_intensidade(self, landmarks):
        """
          Calcula a intensidade geral do movimento
        """
        #print("[DEBUG] Iniciando cálculo de intensidade")
        try:
            # Usa pontos-chave como pulsos e tornozelos
            pontos_chave = [landmarks[i] for i in [15, 16, 29, 30]]

            x_coords = [p.x for p in pontos_chave]
            y_coords = [p.y for p in pontos_chave]

            amplitude_x = max(x_coords) - min(x_coords)
            amplitude_y = max(y_coords) - min(y_coords)

            intensidade = (amplitude_x + amplitude_y) / 2
         #   print(f"[DEBUG] Intensidade calculada: {intensidade}")
            return intensidade

        except Exception as e:
            print(f"[ERROR] Erro ao calcular intensidade: {str(e)}")
            return 0.0

    def _calcular_equilibrio(self, landmarks):
        """
        Calcula o equilíbrio baseado na posição dos ombros
        """
       # print("[DEBUG] Iniciando cálculo de equilíbrio")
        try:
            ombro_dir = landmarks[12]  # Índice do ombro direito
            ombro_esq = landmarks[11]  # Índice do ombro esquerdo
            equilibrio = abs(ombro_dir.y - ombro_esq.y)
        #    print(f"[DEBUG] Equilíbrio calculado: {equilibrio}")
            return equilibrio
        except Exception as e:
            print(f"[ERROR] Erro ao calcular equilíbrio: {str(e)}")
            return 0.0

    def _calcular_movimento_vertical(self, landmarks):
        """
        Calcula o movimento vertical baseado na posição dos quadris
        """
       # print("[DEBUG] Iniciando cálculo de movimento vertical")

        try:
            quadril_dir = landmarks[24]  # Índice do quadril direito
            quadril_esq = landmarks[23]  # Índice do quadril esquerdo
            movimento = abs((quadril_dir.y + quadril_esq.y) / 2 - 0.5)
        #    print(f"[DEBUG] Movimento vertical calculado: {movimento}")
            return movimento
        except Exception as e:
            print(f"[ERROR] Erro ao calcular movimento vertical: {str(e)}")
            return 0.0

    def _calcular_amplitude_movimento(self, landmarks):
        """Calcula a amplitude do movimento baseado na dispersão dos pontos"""
        pontos_chave = [
          self.mp_pose.PoseLandmark.LEFT_WRIST,
            self.mp_pose.PoseLandmark.RIGHT_WRIST,
            self.mp_pose.PoseLandmark.LEFT_ANKLE,
            self.mp_pose.PoseLandmark.RIGHT_ANKLE
       ]

        x_coords = [landmarks[p.value].x for p in pontos_chave]
        y_coords = [landmarks[p.value].y for p in pontos_chave]

        amplitude_x = max(x_coords) - min(x_coords)
        amplitude_y = max(y_coords) - min(y_coords)

        return (amplitude_x + amplitude_y) / 2

    def _avaliar_mudanca_direcao(self, direcao_atual, direcao_anterior):
        """Avalia se houve mudança significativa de direção"""
        if direcao_anterior is None:
            return 0.0
        return 1.0 if direcao_atual != direcao_anterior else 0.0

    def _calcular_intensidade_movimento(self, landmarks, velocidade):
        """Calcula a intensidade geral do movimento"""
        # Combina velocidade com amplitude de movimento
        amplitude = self._calcular_amplitude_movimento(landmarks)
        return (velocidade + amplitude) / 2

    def _calcular_direcao_movimento(self, landmarks_atuais, landmarks_anteriores):
        """Calcula a direção principal do movimento"""
        nariz_atual = landmarks_atuais[self.mp_pose.PoseLandmark.NOSE.value]
        nariz_anterior = landmarks_anteriores[self.mp_pose.PoseLandmark.NOSE.value]

        dx = nariz_atual.x - nariz_anterior.x
        dy = nariz_atual.y - nariz_anterior.y

        if abs(dx) > abs(dy):
            return 'horizontal'
        else:
            return 'vertical'

    def get_statistics(self):
        """
        Gera estatísticas das anomalias detectadas
        """
        stats = {
        'total_anomalias': 0,
        'anomaly_types':self.anomaly_types.copy(),
        "severity_levels": {"baixo": 0, "medio": 0, "alto": 0},
        "time_distribution": {},
      }

        # Processa histórico apenas para severidade e distribuição temporal
        for entry in self.anomaly_history:
            try:
            # Distribuição temporal
                timestamp = entry.get("timestamp", 0)
                time_bucket = int(timestamp / 5) * 5
                stats["time_distribution"][time_bucket] = stats["time_distribution"].get(time_bucket, 0) + 1

                # Severidade baseada no número de anomalias simultâneas no frame
                anomalies_count = sum(1 for v in entry.get("anomalies", {}).values() if v > 0)

                if anomalies_count == 1:
                    stats["severity_levels"]["baixo"] += 1
                elif anomalies_count == 2:
                    stats["severity_levels"]["medio"] += 1
                elif anomalies_count > 2:
                    stats["severity_levels"]["alto"] += 1

            except Exception as e:
                self._logger.log(f" EnhancedAnomalyDetector: Erro ao processar entrada do histórico: {str(e)}")
                continue

        return stats

    ##############################################################
    # analyze_frame - analisa frame de movimentos anômalos
    ##############################################################

    def analyze_frame(self, frame_data):

        try:
            logger.log('[Enhanced][analyze_frame] nova versao [EnhancedAnomalyDetector] ')

            if frame_data is None:
                logger.log('[Enhanced][analyze_frame] frame_data vazio')
                return self.default_anomalies

        # extração de movements do frame_data
            if isinstance(frame_data, dict):
                movements = frame_data.get('movements', {})
            else:
                frame = np.asarray(frame_data, dtype=np.uint8)
                self._logger.log('[Enhanced][analyze_frame] Chamando self._extract_movements')
                movements = self._extract_movements(frame)

        # Modificar o log_anomalies para incluir frame_id
            frame_id = frame_data.get('frame_id', 0) if isinstance(frame_data, dict) else 0

        # Detecta anomalias
            movimento_brusco = self._detect_speed_anomalies(movements)
            tremor = self.anomaly_detector._detect_tremor(movements)
            queda = self.anomaly_detector._detect_fail(movements)
            desequilibrio = self.anomaly_detector._detect_balance_loss(movements)
            colisao = self.anomaly_detector._detect_collision(movements)
            desmaio = self.anomaly_detector._detect_faint(movements)

        # Atualiza contadores
            anomalies = self._get_current_anomalies()

        # Inicia sessão se necessário
            if anomalies['tempo_inicio_sessao'] is None:
                anomalies['tempo_inicio_sessao'] = time.time()

            if movimento_brusco == 1:
                anomalies['movimento_brusco'] = True
                anomalies['total_movimento_brusco'] += 1
                anomalies['sessao_movimento_brusco'] += 1

            if tremor == 1:
                anomalies['tremor'] = True
                anomalies['total_tremor'] += 1
                anomalies['sessao_tremor'] += 1

            if queda == 1:
                anomalies['queda'] = True
                anomalies['total_queda'] += 1
                anomalies['sessao_queda'] += 1

            if desequilibrio == 1:
                anomalies['desequilibrio'] = True
                anomalies['total_desequilibrio'] += 1
                anomalies['sessao_desequilibrio'] += 1

            if colisao == 1:
                anomalies['colisao'] = True
                anomalies['total_colisao'] += 1
                anomalies['sessao_colisao'] += 1

            if desmaio == 1:
                anomalies['desmaio'] = True
                anomalies['total_desmaio'] += 1
                anomalies['sessao_desmaio'] += 1

        # Atualiza métricas gerais
            anomalies['intensidade'] = movements.get('intensidade', 0.0)
            anomalies['confianca'] = movements.get('confianca', 0.0)
            anomalies['movimento_anomalo'] = any([
            movimento_brusco == 1,
            tremor == 1,
            queda == 1,
            desequilibrio == 1,
            colisao == 1,
            desmaio == 1
        ])

            if anomalies['movimento_anomalo']:
                anomalies['total_anomalias'] += 1
                anomalies['total_anomalias_sessao'] += 1

        # Atualiza duração da sessão
            anomalies['duracao_sessao'] = time.time() - anomalies['tempo_inicio_sessao']

        # Log das anomalias detectadas
            self.log_anomalies(anomalies, frame_id)

        # Salva estado atual
            self._save_current_anomalies(anomalies)
            return anomalies

        except Exception as e:
              self._logger.log(f"[EnhancedAnomalyDetector][analyze_frame] Erro na análise: {str(e)}")
              return self.default_anomalies


    def _get_current_anomalies(self):
        """
        Recupera estado atual dos contadores
        """
        if not hasattr(self, '_current_anomalies'):
           self._current_anomalies = self.default_anomalies
        return self._current_anomalies

    def _save_current_anomalies(self, anomalies):
        """
        Salva estado atual dos contadores
        """
        self._current_anomalies = anomalies

    def _detect_speed_anomalies(self, movements):
        """
        Detecta movimentos bruscos considerando todos os tipos de movimento
        Args:
            movements: Dicionário com dados de movimento
          """
        try:
            # Thresholds para cada tipo de movimento
            VELOCIDADE_THRESHOLD = 0.05
            VERTICAL_THRESHOLD = 0.8
            EQUILIBRIO_THRESHOLD = 0.05
            DIRECAO_THRESHOLD = 0.05
            INTENSIDADE_THRESHOLD = 0.8

            # Valores atuais
            velocidade = float(movements['velocidade'])
            movimento_vertical = float(movements['movimento_vertical'])
            equilibrio = float(movements['equilibrio'])
            mudanca_direcao = float(movements['mudanca_direcao'])
            intensidade = float(movements['intensidade'])

            # Debug de cada tipo
            # print(f'[DEBUG] Análise de movimentos:')
            # print(f'Velocidade: {velocidade} (threshold: {VELOCIDADE_THRESHOLD})')
            # print(f'Vertical: {movimento_vertical} (threshold: {VERTICAL_THRESHOLD})')
            # print(f'Equilíbrio: {equilibrio} (threshold: {EQUILIBRIO_THRESHOLD})')
            # print(f'Direção: {mudanca_direcao} (threshold: {DIRECAO_THRESHOLD})')
            # print(f'Intensidade: {intensidade} (threshold: {INTENSIDADE_THRESHOLD})')

          # Detecção para cada tipo
            is_movement_anomaly = (
              velocidade > VELOCIDADE_THRESHOLD or
              movimento_vertical > VERTICAL_THRESHOLD or
              equilibrio > EQUILIBRIO_THRESHOLD or
              mudanca_direcao > DIRECAO_THRESHOLD or
              intensidade > INTENSIDADE_THRESHOLD
          )

            #print(f"[EnhancedAnomalyDetector] Anomalia detectada: {is_movement_anomaly}")
            return 1 if is_movement_anomaly else 0

        except Exception as e:
               print(f"[EnhancedAnomalyDetector] Erro na detecção: {str(e)}")
               return 0

    def _detect_pattern_anomalies(self) -> List[str]:
        """Detecta anomalias baseadas em padrões de movimento"""
        anomalies = []
        try:
            # Analisa últimos N frames
            recent_frames = self.movement_history[-10:]
            landmarks = np.array([frame['landmarks'] for frame in recent_frames])

            # Calcula variância do movimento
            movement_variance = np.var(landmarks, axis=0)
            mean_variance = np.mean(movement_variance)

            if mean_variance > self.thresholds['pattern']:
                anomalies.append(f"Padrão de movimento anormal detectado: {mean_variance:.2f}")

        except Exception as e:
            print(f"Erro ao analisar padrões: {str(e)}")

        return anomalies

    def _detect_direction_anomalies(self) -> List[str]:
        """Detecta mudanças bruscas de direção"""
        anomalies = []
        try:
            # Pega os três últimos frames
            frames = self.movement_history[-3:]
            vectors = []

            # Calcula vetores de movimento
            for i in range(len(frames)-1):
                current = np.array(frames[i]['landmarks'])
                next_frame = np.array(frames[i+1]['landmarks'])
                vectors.append(next_frame - current)

            # Calcula ângulo entre vetores
            dot_product = np.dot(vectors[0].flatten(), vectors[1].flatten())
            magnitudes = np.linalg.norm(vectors[0]) * np.linalg.norm(vectors[1])

            if magnitudes > 0:
                angle = np.arccos(dot_product / magnitudes)
                if angle > self.thresholds['direction']:
                    anomalies.append(f"Mudança brusca de direção detectada: {np.degrees(angle):.2f}°")

        except Exception as e:
            print(f" EnhancedAnomalyDetector: Erro ao analisar direção: {str(e)}")

        return anomalies


    def _record_analysis(self,
                        anomalies: Dict[str, List[str]],
                        timestamp: datetime) -> None:
        """Registra resultados da análise"""
        if any(anomalies.values()):
            self.anomaly_history.append({
                'timestamp': timestamp,
                'anomalies': anomalies
            })

    def get_statistics(self) -> Dict[str, Any]:
        """Retorna estatísticas das análises"""
        total_anomalies = len(self.anomaly_history)

        stats = {
            'total_frames_analyzed': len(self.movement_history),
            'total_anomalias': total_anomalies,
            'anomaly_types': {
                'speed': sum(1 for record in self.anomaly_history
                           if record['anomalies']['speed_anomalies']),
                'pattern': sum(1 for record in self.anomaly_history
                             if record['anomalies']['pattern_anomalies']),
                'direction': sum(1 for record in self.anomaly_history
                               if record['anomalies']['direction_anomalies'])
            }
        }

        return stats

    def reset(self) -> None:
        """Reseta o detector"""
        self.movement_history.clear()
        self.anomaly_history.clear()
        self.last_analysis_time = None

    def update_thresholds(self, new_thresholds: Dict[str, float]) -> None:
        """Atualiza limiares de detecção"""
        self.thresholds.update(new_thresholds)

    def draw_anomalies(self, frame, anomalies):
        """
        Desenha indicadores visuais para as anomalias detectadas no frame
        """

        print("EnhancedAnomalyDetector: draw_anomalies()")

        try:
            # Cria uma cópia do frame para não modificar o original
            annotated_frame = frame.copy()

            # Define parâmetros visuais
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.5
            thickness = 2

            # Posição inicial para texto
            y_position = 30

            # Desenha indicadores para cada tipo de anomalia detectada
            for anomaly_type, value in anomalies.items():
                if value > 0:
                    # Texto em vermelho para anomalias detectadas
                    cv2.putText(
                        annotated_frame,
                        f"ALERTA: {anomaly_type}",
                        (10, y_position),
                        font,
                        font_scale,
                        (0, 0, 255),  # Cor vermelha
                        thickness
                    )
                    y_position += 20

            return annotated_frame

        except Exception as e:
            print(f"Enhanced: Erro ao desenhar anomalias: {str(e)}")
            return frame  # Retorna frame original em caso de erro
```

 - Faz a orquestração dos diversos detectores de atividades, tais como:
    Faces, Expressões Faciais, Movimentos Corporais, Movimentos Anômalos

```python
# 2.5 - Processamento de video - Classe VideoProcessor
import mediapipe as mp
import cv2

mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

class VideoProcessor:

    def __init__(self, detectors):

        self._logger = CustomLogger(
            show_console=  False,
            log_debug= False
           )

        logger.log("[VideoProcessor] Iniciando...")

        self.cap = None
        self.total_frames = 0
        self.current_frame = 0

        self.detectors = detectors
        self.display = DisplayProgressing()
        self.previous_activities = []

         # Instancia objetos
        self.anomaly_detector = AnomalyMovementDetector()
        self.enhanced_anomaly_detector = EnhancedAnomalyDetector()
        self.report_view = ReportView(analysis_results=self.anomaly_detector.anomaly_counts)

        self.default_anomalies = {
            'tremor': False,
            'queda': False,
            'colisao': False,
            'delirio': False,
            'desmaio': False,
            'desequilibrio': False,
            'movimento_brusco': False,
            'total_anomalias': 0,
            'total_anomalias_sessao': 0,
            'tempo_inicio_sessao': None,
            'duracao_sessao': 0,
            # Adicionando os novos contadores
            'sessao_movimento_brusco': 0,
            'sessao_tremor': 0,
            'sessao_queda': 0,
            'sessao_desequilibrio': 0,
            'sessao_colisao': 0,
            'sessao_desmaio': 0,
            'total_movimento_brusco': 0,
            'total_tremor': 0,
            'total_queda': 0,
            'total_desequilibrio': 0,
            'total_colisao': 0,
            'total_desmaio': 0
         }

        # Inicializa contadores de anomalias com os mesmos tipos
        self.anomaly_counts = self.default_anomalies.copy()

        # Inicialização do MediaPipe
        self.mp_pose = mp.solutions.pose
        self.pose = self.mp_pose.Pose(
            min_detection_confidence=0.5,
            min_tracking_confidence=0.5,
            model_complexity=1
        )

        # Inicializa o dicionário de tempos de processamento
        self.processing_times = {
            "Detecção de Faces": 0.0,
            "Detecção de Emoções": 0.0,
            "Detecção de Atividades": 0.0,
            "Detecção de Movimentos": 0.0,
            "Detecção de Anomalias": 0.0,
            "Gravação do Frame": 0.0,
            "Desenho de Emoções": 0.0,
        }

          # Inicializa os resultados da análise
        self.analysis_results = {
            "frames_analyzed": 0,
            "faces_detected": 0,
            "emotion_counts": {},
            "anomalies_detected": 0,
            "total_movements": 0,
            "output_video_path": "",
            "previous_activities": [],

            "body_movements":
                {
                    "braco_levantado": 0,
                    "mao_gesticulando": 0,
                    "caminhando": 0,
                    "sentado": 0,
                    "em_pe": 0,
                    "deitado": 0,
                    "aperto_maos":0
                    # outros tipos, 'total_estereotipia_oral(movimentos da boca)
                    },
            "anomaly_movements":
               {
                  "tremor": 0,
                  "queda": 0,
                  "movimento_brusco"  : 0,
                  "desequilibrio": 0,
                  "colisao": 0,
                  "desmaio": 0,
                  "movimento_anomalo": 0
                  },
             "anomalies":
               {
                  "total": 0,
                  "by_type": {}
              }
        }

    def process_video(self, video_path, output_path, output_file, frame_skip=10):
        try:
            # Abre o vídeo
            cap = cv2.VideoCapture(video_path)
            if not cap.isOpened():
                raise Exception(f"Erro ao abrir o vídeo: {video_path}")

            # Obtém informações do vídeo
            video_info = {
                'total_frames': int(cap.get(cv2.CAP_PROP_FRAME_COUNT)),
                'frame_width': int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
                'frame_height': int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)),
                'fps': int(cap.get(cv2.CAP_PROP_FPS))
            }

            # total de sps e frames do video
            fps = int(cap.get(cv2.CAP_PROP_FPS))
            self.total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

            # Cria diretório de saída se não existir
            os.makedirs(output_path, exist_ok=True)

            #print("VideoProcessor(): output_file" + output_file)
            output_video_path = os.path.join(output_path, output_file)

            #print(f'"VideoProcessor(): output_video_path", output_video_path')
            self.analysis_results['output_video_path'] = output_video_path

            # Configura o writer do vídeo de saída
            out = cv2.VideoWriter(
                output_video_path,
                cv2.VideoWriter_fourcc(*'mp4v'),
                video_info['fps'],
                (video_info['frame_width'], video_info['frame_height'])
            )

            if not out.isOpened():
                raise Exception(f"Erro ao criar arquivo de saída: {output_video_path}")

            previous_activities = []
            processed_count = 0
            frame_count = 0

            ######################################################################
            # loop para leitura do video a cada 'n' frames, conforme frame_skip
            ######################################################################

            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                frame_count += 1
                if frame_count % frame_skip != 0:
                    continue

                ######################################################################
                # 1-  Detecção de expressões faciais
                ######################################################################

                # 1.1- Detecção de faces
                start_time = time.time()
                frame, faces = self.detectors.detect_faces(frame)
                self.processing_times['Detecção de Faces'] = (time.time() - start_time) * 1000
                self.analysis_results['faces_detected'] += len(faces)

                # 1.2 - Detecção de emoções
                start_time = time.time()
                emotions = self.detectors.analyze_emotions(frame)
                self.processing_times['Detecção de Emoções'] = (time.time() - start_time) * 1000

                # 1.3 - Escreve a descrição das emoções no frame
                start_time = time.time()
                frame = self.detectors.draw_emotions(frame, emotions, faces)
                self.processing_times['Desenho de Emoções'] = (time.time() - start_time) * 1000

                ######################################################################
                # 2 - Detecção de movimentos corporais
                ######################################################################
                # 2.1 - Detecção de movimentos usando o movement_detector
                start_time = time.time()
                movements = self.detectors.pose_analyzer.detect_movements(frame)
                self.processing_times['Detecção de Movimentos'] = (time.time() - start_time) * 1000

                # 2.2 - Detecção de movimentos corporais
                logger.log('[VidelProcesso][pose_analyzer.draw_movements]...')
                frame = self.detectors.pose_analyzer.draw_movements(frame, movements)
                self.processing_times['Detecção de Movimentos'] = (time.time() - start_time) * 1000

                for movement, value in movements.items():
                   if movement != 'movimento_anomalo' and value > 0:
                      self.analysis_results['body_movements'][movement] = self.analysis_results['body_movements'].get(movement, 0) + 1

                # 2.3 - Atualiza total de movimentos
                self.analysis_results['total_movements'] = sum(self.analysis_results['body_movements'].values())

                ######################################################################
                # 3 - Detecção de movimentos anômalos com EnhancedAnomalyDetector
                ######################################################################

                movements = self.enhanced_anomaly_detector._extract_movements(frame)

                frame_data = {
                  'movements': movements,
                  'timestamp': frame_count / fps,
                  'frame_id': frame_count
                      }

                # Análise de anomalias
                start_time = time.time()

                anomalies = self.enhanced_anomaly_detector.analyze_frame(frame_data)
                # if not anomalies:
                #    print("[VideoProcessor][process_video]: Nenhuma anomalia detectada")
                # else:
                #    print("[VideoProcessor][process_video]: Erro ao processar anomalias - nenhum dado retornado")

                self.processing_times['Detecção de Anomalias'] = (time.time() - start_time) * 1000

                # Desenha as anomalias no frame
                frame = self.report_view.draw_anomalous_movements(frame, anomalies)

                # Atualiza contadores de anomalia
                if isinstance(anomalies, dict):
                   for anomaly_type, is_detected in anomalies.items():
                       if is_detected:
                       # Incrementa o contador específico do tipo de anomalia
                          if anomaly_type not in self.analysis_results["anomaly_movements"]:
                             self.analysis_results["anomaly_movements"][anomaly_type] = 0
                          self.analysis_results["anomaly_movements"][anomaly_type] += 1

                          # Incrementa o contador total de anomalias
                          self.analysis_results["anomalies_detected"] += 1

                   # Coleta estatísticas periódicas
                if processed_count % 100 == 0:
                    stats = self.enhanced_anomaly_detector.get_statistics()
                    # print(f"Estatísticas de anomalias (frame {frame_count})")
                    # print(f"Total de anomalias detectadas: {self.analysis_results['anomalies_detected']}")
                    # print(f"Tipos de anomalias: {self.analysis_results['anomaly_movements']}")

                ### --------------------- Fim Anomalias ---------------------------------------------------

                ################################################
                # 4 - Atualização resultado de análise
                ################################################
                self._update_analysis(faces,
                                    emotions,
                                    movements,
                                    anomalies)

                ################################################
                # 5 - Exibição progresso do processamento na tela
                ################################################
                self._display_progress(frame_count, video_info["total_frames"])

                processed_count += 1

                ################################################***
                # 7 - Exibição progresso do processamento no video
                ###############################################***
                if frame_count % 10 == 0:

                  # Adiciona a barra de progresso ao frame
                   frame = self.display.desenha_barra_progresso(
                            frame,
                            frame_count,
                            video_info["total_frames"],
                            frame_skip)

                ################################################
                # 6 - Gravação do frame processado
                ################################################
                out.write(frame)

            #
            # Libera recursos
            cap.release()
            out.release()

            logger.log("\nProcessamento concluído. Arquivo salvo em:" + output_video_path)

            frames_processados = f'{processed_count}/{frame_count}'
            print(frames_processados)

            return self.analysis_results

        except Exception as e:
            print(f"Erro no processamento do frame {frame_count}: {str(e)}")
            print(f"Erro durante processamento: {str(e)}")
            if "out" in locals():
                out.release()
            if "cap" in locals():
                cap.release()
            raise

    # VideoProcessor
    def _update_analysis(self, faces, emotions, movements, anomalies):
        """Atualiza os resultados da análise"""
        # Atualiza contadores de faces e frames
        #print('[VideoProcessor][_update_analysis]')
        self.analysis_results["faces_detected"] += len(faces)
        self.analysis_results["frames_analyzed"] += 1

    # Atualiza contadores de emoções
        for emotion in emotions:
            self.analysis_results["emotion_counts"][emotion] = self.analysis_results["emotion_counts"].get(emotion, 0) + 1

        # Atualiza contadores de movimento
        for movement, value in movements.items():
            if value > 0:  # Só incrementa se houver detecção
                self.analysis_results["body_movements"][movement] = self.analysis_results["body_movements"].get(movement, 0) + 1

        # Atualiza anomalias - com tratamento mais robusto
        if isinstance(anomalies, dict) and anomalies is not None :
            for anomaly, value in anomalies.items():
                if value > 0:
                    # Atualiza anomaly_movements
                    if anomaly in self.analysis_results["anomaly_movements"]:
                        self.analysis_results["anomaly_movements"][anomaly] += 1
                    else:
                        # Cria nova entrada se não existir
                        self.analysis_results["anomaly_movements"][anomaly] = 1

                # Atualiza contadores em anomalies.by_type
                if anomaly not in self.analysis_results["anomalies"]["by_type"]:
                    self.analysis_results["anomalies"]["by_type"][anomaly] = 0
                self.analysis_results["anomalies"]["by_type"][anomaly] += 1
                self.analysis_results["anomalies"]["total"] += 1

        # Atualiza totais
        self.analysis_results["total_movements"] = sum(self.analysis_results["body_movements"].values())
        self.analysis_results["total_anomalies"] = sum(self.analysis_results["anomaly_movements"].values())

    # VideoProcessor
    def _display_progress(self, current_frame, total_frames):
        """Atualiza o display de progresso"""
        if current_frame % 10 == 0:
                self.display.display_processing_status(
                current=current_frame,
                total=total_frames,
                frames_analyzed=self.analysis_results["frames_analyzed"],
                faces_detected=self.analysis_results["faces_detected"],
                emotions_count=len(self.analysis_results["emotion_counts"]),
                emotion_data=self.analysis_results["emotion_counts"],
                body_data=self.analysis_results["body_movements"],
                anomaly_data=self.analysis_results['anomaly_movements'],
                processing_time=sum(self.processing_times.values()),
            )

    # VideoProcessor
    def _get_video_info(self, cap):
        """Obtém informações do vídeo"""
        return {
            "total_frames": int(cap.get(cv2.CAP_PROP_FRAME_COUNT)),
            "frame_width": int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
            "frame_height": int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)),
            "fps": int(cap.get(cv2.CAP_PROP_FPS)),
        }
```

```python
# 3.1 - Armazenar e organizar os dados do relatório - Classe ReportData
class ReportData:
    """
    Classe para armazenar e organizar os dados do relatório
    """
    def __init__(self, analysis_results):
        self.analysis_results = analysis_results
        self.metrics_table = None
        self.emotions_table = None
        self.body_movements_table = None
        self.body_anomalies_table = None
        self.file_info_table = None
        self.summary = None
        self._prepare_data()

        logger.log('Chamando ReportData...')

    def _prepare_data(self):
        """Prepara todos os dados do relatório"""
        try:
            self._prepare_metrics_table()         # Prepara tabela de métricas
            self._prepare_emotions_table()        # Prepara tabela de emoções
            self._prepare_movements_table()       # Prepara tabela de movimentos corporais
            self._prepare_anomalies_table()       # Prepara tabela de movimentos anômalos
            self._prepare_file_info_table()       # Prepara tabela de informações dos arquivos
            self._generate_summary()              # Gera o resumo

        except Exception as e:
            print(f"_prepare_data: Erro ao preparar dados do relatório: {str(e)}")
            raise

     # ReportData
    def _prepare_anomalies_table(self):
        """Prepara tabela de movimentos anômalos"""
        try:
            if "anomaly_movements" in self.analysis_results:
                anomalies = self.analysis_results["anomaly_movements"]
                total_anomalies = sum(anomalies.values())

                anomalies_data = {"Tipo": [], "Quantidade": [], "Porcentagem (%)": []}

            # Labels em português
                anomaly_labels = {
                    "tremor": "Tremor",
                    "queda": "Queda",
                    "movimento_brusco": "Movimento Brusco",
                    "desequilibrio": "Desequilibrio",
                    "colisao": "Colisao",
                    "desmaio": "Desmaio",
            }

            # Só inclui anomalias detectadas
                for anomaly, count in anomalies.items():
                    if count > 0:
                      anomalies_data["Tipo"].append(anomaly_labels.get(anomaly, anomaly))
                      anomalies_data["Quantidade"].append(count)
                      percentage = (count / total_anomalies * 100) if total_anomalies > 0 else 0
                      anomalies_data["Porcentagem (%)"].append(round(percentage, 2))

                 # Criar DataFrame e ordenar por quantidade
                self.anomalies_table = pd.DataFrame(anomalies_data)
                self.anomalies_table = self.anomalies_table.sort_values("Quantidade", ascending=False).reset_index(drop=True)

            else:
                 print("Dados de anomalias não encontrados")

        except Exception as e:
               print(f"Erro ao preparar tabela de anomalias: {str(e)}")

    def _prepare_movements_table(self):
        """Prepara tabela de movimentos corporais"""
        try:

            if 'body_movements' in self.analysis_results:
                movements = self.analysis_results['body_movements']
                total_movements = sum(movements.values())

                movements_data = {
                   'Movimento': [],
                   'Quantidade': [],
                   'Porcentagem (%)': []
               }

            # Definindo o dicionário movement_labels antes do loop
                movement_labels = {
                    'braco_levantado': 'Braço Levantado',
                    'mao_gesticulando': 'Mão Gesticulando',
                    'caminhando': 'Caminhando',
                    'sentado': 'Sentado',
                    'em_pe': 'Em Pé',
                    'deitado': 'Deitado',
                    "aperto_maos": 'Aberto Maos'
              }

             # Só inclui movimentos detectado
            for movement, count in movements.items():
                if count > 0:  # Só inclui movimentos detectados
                   movements_data['Movimento'].append(movement_labels.get(movement, movement))
                   movements_data['Quantidade'].append(count)
                   percentage = (count / total_movements * 100) if total_movements > 0 else 0
                   movements_data['Porcentagem (%)'].append(round(percentage, 2))

            # Criar DataFrame e ordenar por quantidade
            self.body_movements_table = pd.DataFrame(movements_data)
            self.body_movements_table = self.body_movements_table.sort_values(
                                        'Quantidade', ascending=False
           ).reset_index(drop=True)

        except Exception as e:
               print(f"Erro ao preparar tabela de movimentos: {str(e)}")

    def _generate_summary(report_data):  # Método público (sem underline)
        """
        Gera um resumo do relatório
        Args:
            report_data: Instância de ReportData
        Returns:
            str: Texto do resumo
        """
        try:
            metrics = report_data.metrics_table
            emotions = report_data.emotions_table
            movements = report_data.body_movements_table


            summary_parts = [
                "Resumo da Análise:",
                f"- Frames analisados: {metrics.iloc[0]['Valor']}",
                f"- Faces detectadas: {metrics.iloc[1]['Valor']}",
                f"- Anomalias detectadas: {metrics.iloc[2]['Valor']}",
                f"- Novimentos corporais: {metrics.iloc[3]['Valor']}",
              ]

            if not emotions.empty:
                dominant_emotion = emotions.iloc[0]
                summary_parts.append(
                    f"- Emocao predominante: {dominant_emotion['Emocao']} "
                    f"({dominant_emotion['Porcentagem (%)']}%)"
                )

            for _, row in movements.iterrows():
                summary_parts.append(f"- {row['Movimento']}: {row['Quantidade']}")

            return "\n".join(summary_parts)

        except Exception as e:
            print(f"_prepare_dataErro ao gerar resumo: {str(e)}")
            return f"_prepare_dataErro ao gerar resumo:{str(e)}"

    def _prepare_file_info_table(self):
        """Prepara tabela de informações dos arquivos"""
        self.file_info_table = pd.DataFrame({
            'Tipo': ['Vídeo Analisado', 'Relatório'],
            'Caminho': [
                self.analysis_results['output_video_path'],
                os.path.join(
                    os.path.dirname(str(self.analysis_results['output_video_path'])),
                    'Relatorio_Analise_Video.txt'
                )
            ]
        })

    def _prepare_emotions_table(self):
        """Prepara tabela de emoções"""
        frames = self.analysis_results['frames_analyzed']
        emotions_data = {
            'Emocao': [],
            'Contagem': [],
            'Porcentagem (%)': []
        }

        for emotion, count in self.analysis_results['emotion_counts'].items():
            count = int(count)  # Converter contagem para inteiro
            emotions_data['Emocao'].append(emotion)
            emotions_data['Contagem'].append(count)
            percentage = round((count / frames * 100), 2) if frames > 0 else 0
            emotions_data['Porcentagem (%)'].append(percentage)

        self.emotions_table = pd.DataFrame(emotions_data)
        self.emotions_table = self.emotions_table.sort_values(
            'Contagem', ascending=False
        ).reset_index(drop=True)

    def _prepare_metrics_table(self):
        """Prepara tabela de métricas gerais"""

        frames = int(self.analysis_results['frames_analyzed'])
        faces = int(self.analysis_results['faces_detected'])
        anomalies = int(self.analysis_results['anomalies_detected'])

        # Corrigindo o cálculo de body_movements
        movements_dict = self.analysis_results['body_movements']
        total_movements = sum(movements_dict.values()) if isinstance(movements_dict, dict) else 0

        #body_movements = int(self.analysis_results.get('body_movements', 0))  # Usando .get() com valor default

        # Calcular taxa de anomalias
        anomaly_rate = int((anomalies / frames * 100)) if frames > 0 else 0

        # Aqui a metrics_table é preenchida
        self.metrics_table = pd.DataFrame({
            'Métrica': [
                'Frames Analisados',
                'Faces Detectadas',
                'Anomalias Detectadas',
                'Movimentos Corporais',
                'Taxa de Anomalias (%)',
                'Total Movimentos Anômalos'
            ],
            'Valor': [
                frames,
                faces,
                anomalies,
                total_movements,
                anomaly_rate,
                anomalies
            ]
        })
```

### 3.2 - Geração e salvamento do relatório -  Classe ReportGenerator



```python
# 3.2 - Geração e salvamento do relatório - Classe ReportGenerator

class ReportGenerator:
   """
   Classe responsável pela geração e salvamento do relatório
   """
   @staticmethod
   def generate_report(analysis_results):
       """
       Gera os dados do relatório
       Args:
           analysis_results: Dicionário com os resultados da análise
       Returns:
           ReportData: Instância com os dados formatados
       """
       try:

           logger.log('Chamando ReportGenerator...')

           # Verifica campos obrigatórios
           required_fields = ['frames_analyzed',
                              'faces_detected',
                              'emotion_counts',
                              'body_movements',
                              'anomalies_detected']

           for field in required_fields:
               if field not in analysis_results:
                   #print(f"Campo obrigatório ausente: {field}")
                   return None

           data_report = ReportData(analysis_results)
           return data_report

       except Exception as e:

           print(f"ReportGenerator: Erro ao gerar relatório: {e}")
           return None

   @staticmethod
   def generate_summary(report_data):
       """
       Gera um resumo do relatório
       Args:
           report_data: Instância de ReportData
       Returns:
           str: Texto do resumo
       """
       try:
           metrics = report_data.metrics_table
           emotions = report_data.emotions_table
           movements = report_data.body_movements_table
           anomalies = report_data.anomalies_table

           summary_parts = [
               "Resumo da Análise:",
               f"- Frames analisados: {metrics.iloc[0]['Valor']}",
               f"- Faces detectadas: {metrics.iloc[1]['Valor']}",
               f"- Movimentos detectados: {metrics.iloc[2]['Valor']}",
               f"- Anomalias detectadas: {metrics.iloc[3]['Valor']}",
               "Movimentos Corporais:"
           ]

           # Adiciona movimentos ao resumo
           if not movements.empty:
               for _, row in movements.iterrows():
                   summary_parts.append(
                       f"- {row['Movimento']}: {row['Quantidade']} ({row['Porcentagem (%)']}%)"
                   )

           # Adiciona emoções ao resumo
           if not emotions.empty:
               summary_parts.append("\nEmoções Detectadas:")
               for _, row in emotions.iterrows():
                   if row['Quantidade'] > 0:
                       summary_parts.append(
                           f"- {row['Emocao']}: {row['Quantidade']} ({row['Porcentagem (%)']}%)"
                       )


            # Adiciona anomalias ao resumo
           if not anomalies.empty:
               summary_parts.append("\nMovimentos Anômalos Detectados:")
               for _, row in anomalies.iterrows():
                   if row['Quantidade'] > 0:
                      summary_parts.append(
                          f"- {row['Tipo']}: {row['Quantidade']} ({row['Porcentagem (%)']}%)"
                    )

           return "\n".join(summary_parts)

       except Exception as e:
           print(f"generate_summary(): An unexpected error occurred: {e}")

           return f"Erro ao gerar resumo: {str(e)}"

   @staticmethod
   def validate_report_data(report_data):
       """
       Valida os dados do relatório
       Args:
           report_data: Instância de ReportData
       Returns:
           bool: True se válido, False caso contrário
       """
       try:
           # Verifica se as tabelas principais existem
           if report_data.metrics_table is None:
               print("Erro: Tabela de métricas não encontrada")
               return False

           if report_data.emotions_table is None:
               print("Erro: Tabela de emoções não encontrada")
               return False

           if report_data.file_info_table is None:
               print("Erro: Tabela de informações de arquivo não encontrada")
               return False

           if report_data.body_movements_table is None:
               print("Erro: Tabela de movimentos não encontrada")
               return False

           # Verifica se as tabelas têm dados
           if report_data.metrics_table.empty:
               print("Erro: Tabela de métricas vazia")
               return False

           return True

       except Exception as e:
           print(f"Erro na validação dos dados: {str(e)}")
           return False

   @staticmethod
   def save_report(report_data, report_path):
       """
       Salva o relatório em arquivo
       Args:
           report_data: Instância de ReportData
           report_path: Caminho onde salvar o relatório
       Returns:
           bool: True se salvou com sucesso, False caso contrário
       """
       try:
           if not isinstance(report_data, ReportData):
               raise ValueError("report_data deve ser uma instância de ReportData")

           os.makedirs(os.path.dirname(report_path), exist_ok=True)

           report_content = [
						      "="*50, "RELATÓRIO DE ANÁLISE DE VÍDEO",
                  "="*50,	f"\nData: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}",
						      "\n1. MÉTRICAS GERAIS:","-"*30,
            			report_data.body_movements_table.to_string(),	"\n3. ANÁLISE DE EMOÇÕES:",	"-"*30,
						      report_data.metrics_table.to_string(),        "\n2. ANÁLISE DE MOVIMENTOS:","-"*30,
                  report_data.emotions_table.to_string(),  			"\n4. MOVIMENTOS ANÔMALOS:","-"*30,
						      report_data.anomalies_table.to_string(),   		"\n5. INFORMAÇÕES DOS ARQUIVOS:",	"-"*30,
						      report_data.file_info_table.to_string(),			"\nRESUMO:","-"*30,
						      report_data.summary if report_data.summary else "Resumo não disponível","\n" + "="*50
					    ]

           with open(report_path, 'w', encoding='utf-8') as f:
               f.write("\n".join(report_content))

           return True

       except Exception as e:
           print(f"Erro ao salvar relatório: {str(e)}")
           return False
```

### 3.3 - Visualização dos dados na tela e video -  Classe ReportView   

```python
# 3.3 - Classe ReportView - Visualização dos dados na tela e video

class ReportView:
    """
    Classe responsável pela visualização dos dados em formato tabular
    """
    def __init__(self, analysis_results):
        self.results = analysis_results

        self._logger = CustomLogger(
                  show_console= LOG_SHOW_CONSOLE,
                  log_debug=  LOG_DEBUG
        )

        self._logger.log('Chamando ReportView...')

        self.anomaly_frame_color = (220, 200, 200)  # Cinza claro

        self.anomaly_counters = {
        'tremor': 0,
        'queda': 0,
        'movimento_brusco': 0,
        'desequilibrio': 0,
        'colisao': 0,
        'desmaio': 0
        # outros tipos, se tiver
        }

        # Cores para os diferentes tipos de anomalias
        self.anomaly_colors = {
            'tremor': (0, 165, 255),       # Laranja
            'queda': (0, 0, 255),          # Vermelho
            'desmaio': (255, 0, 0),        # Azul
            'colisao': (0, 140, 255),      # Laranja escuro
            'desequilibrio': (128, 0, 128), # Roxo
            'total': (50, 50, 50),         # Cinza escuro
        }

    def format_table_style(self, df, gradient_column=None):
        """Aplica formatação personalizada às tabelas"""
        # Formatação básica
        style = df.style.set_properties(**{
            'background-color': '#f5f5f5',
            'border-color': 'black',
            'border-style': 'solid',
            'border-width': '1px',
            'text-align': 'center'
        })

        # Formatação específica para porcentagem
        if 'Porcentagem (%)' in df.columns:
            style = style.format({
                'Porcentagem (%)': '{:.2f}'
            })

        if gradient_column and gradient_column in df.columns:
            style = style.background_gradient(subset=[gradient_column], cmap='Blues')

        return style

    def display_metrics_table(self, metrics_df):
        """Exibe tabela de métricas gerais"""
        print("\n=== Métricas Gerais ===")
        display(self.format_table_style(metrics_df))

    def display_emotions_table(self, emotions_df):
        """Exibe tabela de emoções"""
        print("\n=== Análise de Emoções ===")
        display(self.format_table_style(emotions_df, 'Porcentagem (%)'))

    def display_movements_table(self, movements_df):
        print("\n=== Movimentos Corporais ===")
        display(self.format_table_style(movements_df))

    def display_anomalies_table(self, anomalies_df):
        """Exibe tabela de movimentos anômalos"""
        print("\n=== Movimentos Anômalos ===")
        display(self.format_table_style(anomalies_df, 'Porcentagem (%)'))

    def display_file_info_table(self, file_info_df):
        """Exibe tabela de informações dos arquivos"""
        print("\n=== Informações dos Arquivos ===")
        display(self.format_table_style(file_info_df))

    def display_full_report(self, report_data):
        """Exibe o relatório completo em formato visual"""
        self.display_metrics_table(report_data.metrics_table)
        self.display_emotions_table(report_data.emotions_table)
        self.display_movements_table(report_data.body_movements_table)
        self.display_anomalies_table(report_data.anomalies_table)
        self.display_file_info_table(report_data.file_info_table)
        print("\nResumo:")
        print(report_data.summary)


    # Desenha contadores de movimentos anomalos detecados no video
    def update_anomaly_counters(self, anomalies):
        if anomalies:
           for anomaly_type, value in anomalies.items():
                if anomaly_type in self.anomaly_counters:
                # Converte valor booleano para número (0 ou 1)
                   if isinstance(value, bool):
                      self.anomaly_counters[anomaly_type] += 1 if value else 0
                   else:
                      self.anomaly_counters[anomaly_type] = value

    # Desenha o quadro de exibiçao de movimentos anômalos
    def draw_anomalous_movements(self, frame, anomalies):
        #Atualiza os contadores antes de desenhar
        self.update_anomaly_counters(anomalies)

        frame_width = frame.shape[1]
        x_start = frame_width - 320

        #Desenha o quadro principal com a cor de fundo cinza claro
        cv2.rectangle(frame, (x_start, 0), (frame_width, 230),
                     self.anomaly_frame_color, -1)

        #Adiciona uma linha sutil na borda
        cv2.line(frame, (x_start, 0), (x_start, 230),
                (180, 180, 180), 1)

        #Título
        titulo_color = (0, 102, 204)  # Laranja escuro/avermelhado
        cv2.putText(frame, 'Movimentos Anomalos', (x_start + 16, 31),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (50, 50, 50), 2)  # sombra
        cv2.putText(frame, 'Movimentos Anomalos', (x_start + 15, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, titulo_color, 2)

        #Exibe os contadores de anomalias
        y_pos = 60
        for anomaly_type in ['tremor', 'queda', 'desmaio', 'colisao', 'desequilibrio']:
            count = self.anomaly_counters.get(anomaly_type, 0)
            color = self.anomaly_colors.get(anomaly_type, (50, 50, 50))
            text = f'{anomaly_type.capitalize()}: {count}'
            cv2.putText(frame, text, (x_start + 15, y_pos),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
            y_pos += 30

        #Total de anomalias
        total_anomalias = (self.anomaly_counters['tremor'] +
                          self.anomaly_counters['queda'] +
                          self.anomaly_counters['desmaio'] +
                          self.anomaly_counters['colisao'] +
                          self.anomaly_counters['desequilibrio'])

        cv2.putText(frame, f'Total: {total_anomalias}',
                    (x_start + 15, y_pos),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, self.anomaly_colors['total'], 2)

          #Indicador de anomalias ativas
        if total_anomalias > 0:
            indicator_color = (0, 0, 255)  # Vermelho
            cv2.circle(frame, (x_start + 10, 15), 5, indicator_color, -1)

        return frame

```

### 3.5 - Exibição do progresso do Processamento - Classe DisplayProgressing

```python
# 3.5 - Exibição e Status do Processamento - Classe DisplayProgressing

import pandas as pd
import time
from IPython.display import clear_output, display

class DisplayProgressing:
    def __init__(self):

        self._logger = CustomLogger(
                       show_console= LOG_SHOW_CONSOLE,
                       log_debug=  LOG_DEBUG    )

        self.styles = [
            dict(selector="", props=[
                ("border-collapse", "collapse"),
                ("width", "95%"),
                ("margin", "10px 0")
            ]),
            dict(selector="th", props=[
                ("text-align", "left"),
                ("background-color", "#f3f3f3"),
                ("padding", "4px 8px"),
                ("border", "1px solid #ddd"),
                ("font-size", "11px"),
                ("width", "22%")
            ]),
            dict(selector="td", props=[
                ("text-align", "left"),
                ("padding", "4px 8px"),
                ("border", "1px solid #ddd"),
                ("font-size", "11px"),
                ("white-space", "nowrap")
            ])
        ]
        self.start_time = time.time()

    def draw_progress_bar(self, current, total):
        try:
            percentual = (current / total) * 100
            msg_progresso = f"{percentual:.2f}% ({current}/{total})"

            bar_length = 30  # Tamanho fixo da barra
            filled_length = int(bar_length * current / total)

        # Garante que filled_length não exceda bar_length
            filled_length = min(filled_length, bar_length)

            bar = '█' * filled_length + '-' * (bar_length - filled_length)
            return f'|{bar}| {msg_progresso}'

        except Exception as e:
              return "|--|"

    def ms_to_seconds(self, ms):
        return f"{(ms/1000):.2f}s"

    def create_status_table(self, frames, faces, emotions, exec_time, total_time, emotion_data, body_data=None, anomaly_data=None):
        try:
            self._logger.log('[DisplayProgressing][create_status_table] - Inicio')

            # Calcular totais
            total_movements = sum(body_data.values()) if body_data else 0
            total_anomalies = sum(anomaly_data.values()) if anomaly_data else 0
            total_expressoes = sum(emotion_data.values()) if emotion_data else 0

             # 1a coluna - Processamento
            processamento = [
                f"Frames processados: {frames}",
                f"Faces detectadas: {faces}",
                "Expressoes faciais:",
                f"- Tipos emocoes: {len(emotion_data) if emotion_data else 0}",
                f"- Total expressoes: {emotions}",
                f"Movimentos corporais: {sum(body_data.values()) if body_data else 0}",
                f"Movimentos anômalos: {body_data.get('movimento_anomalo', 0) if body_data else 0}",  # Usa o valor de movimentos_anomalo
               "Tempo decorrido:",
                f"- Execucao: {exec_time:.2f}s",
                f"- Total: {total_time:.2f}s"
             ]

           # Lista de movimentos corporais
            movimento_tipos = [
              'braco_levantado',
              'mao_gesticulando',
              'caminhando',
              'sentado',
              'em_pe',
              'aperto_maos'
            ]

            anomalos_tipos = [
              'tremor',
              'queda',
              'movimento_brusco',
              'desequilibrio',
              'colisao',
              'desmaio'
            ]

            tipos_anomalias = [
              'tremor',
              'queda',
              'colisao',
              'delirio',
              'desmaio',
              'desequilibrio',
              'movimento_brusco'
               ]

            expressoes = []
            if emotion_data:
               for emotion, count in emotion_data.items():
                   if count > 0:
                      expressoes.append(f"{emotion}: {count}")

            movimentos_corporais = []
            if body_data:
               for movimento in movimento_tipos:
                   count = body_data.get(movimento, 0)
                   if count > 0:
                     movimentos_corporais.append(f"{movimento}: {count}")

            movimentos_anomalos = []

            for anomalia in tipos_anomalias:
                count = anomaly_data.get(anomalia, 0)
                if count >= 0:
                   movimentos_anomalos.append(f"{anomalia}: {count}")

            # Garantir alinhamento
            # Encontrar o maior comprimento entre todas as colunas
            max_len = max(len(processamento), len(expressoes), len(movimentos_corporais), len(movimentos_anomalos))

        # Padronizar o comprimento de todas as colunas
            processamento = processamento + [''] * (max_len - len(processamento))
            expressoes = expressoes + [''] * (max_len - len(expressoes))
            movimentos_corporais = movimentos_corporais + [''] * (max_len - len(movimentos_corporais))
            movimentos_anomalos = movimentos_anomalos + [''] * (max_len - len(movimentos_anomalos))

            # Criar DataFrame com novo título para coluna de movimentos
            df = pd.DataFrame({
                'Processamento': processamento,
                'Expressoes faciais': expressoes,
                'Movimentos Corporais': movimentos_corporais,
                'Movimentos Anômalos':  movimentos_anomalos
            })

            return df

        except Exception as e:
               print(f"Erro ao criar tabela: {str(e)}")
               return pd.DataFrame()

    def display_processing_status(self, current, total,
                                  frames_analyzed=0,
                                  faces_detected=0,
                                  emotions_count=0,
                                  emotion_data=None,
                                  body_data=None,
                                  anomaly_data=None,
                                  processing_time=0):
        try:
            clear_output(wait=True)   # limpar tela se true

            print(f'Progresso: {self.create_progress_bar(current, total)} Qtde Frames Captura:{FRAME_COUNT_SKIP}')

            total_time = (time.time() - self.start_time) * 1000

            df = self.create_status_table(
                  frames=frames_analyzed,
                  faces=faces_detected,
                  emotions=emotions_count,
                  exec_time=processing_time,
                  total_time=total_time,
                  emotion_data=emotion_data or {},
                  body_data=body_data or {},
                  anomaly_data=anomaly_data or {}
             )

            if not df.empty:
                styled_df = df.style.set_table_styles(self.styles).hide(axis='index')
                display(styled_df)
            else:
                print("display_processing_status: Sem informações disponíveis")

        except Exception as e:
            print(f"display_processing_status: Erro ao exibir status: {str(e)}")


    def create_progress_bar(self, current, total):
        try:
            percent = int((current / total) * 100)
            bar_length = 30
            filled_length = int(bar_length * current / total)
            bar = '█' * filled_length + '-' * (bar_length - filled_length)
            return f'|{bar}| {percent}% ({current}/{total})'
        except Exception as e:
            return "|--|"

    def desenha_barra_progresso(self, frame, current_frame, total_frames,frame_count_skip=1):

        try:
        # Verifica se frame é válido
            if frame is None or not hasattr(frame, 'shape'):
                logger.log("Frame inválido para desenhar barra de progresso")
                return frame

            if len(frame.shape) < 2:
                logger.log("Frame não tem dimensões válidas")
                return frame

            frame_height, frame_width = frame.shape[:2]

        # Posicionamento e dimensões da barra - convertendo para int
            margin = int(20)
            bar_height = int(30)
            bar_width = int(frame_width - (2 * margin))
            y_pos = int(frame_height - margin - bar_height)

        # Retângulo base - garantindo coordenadas inteiras
            cv2.rectangle(frame,
                     (int(margin), int(y_pos)),
                     (int(margin + bar_width), int(y_pos + bar_height)),
                     (200, 200, 200),
                     -1)

        # Calcula percentual e largura do progresso
            percentual = int((current_frame / total_frames) * 100)
            progress_width = int(bar_width * (percentual/100))

        # Desenha barra de progresso - garantindo coordenadas inteiras
            cv2.rectangle(frame,
                     (int(margin), int(y_pos)),
                     (int(margin + progress_width), int(y_pos + bar_height)),
                     (255, 128, 0),
                     -1)

        # Texto de progresso
            msg_progresso = f"{percentual:.2f}% ({current_frame}/{total_frames}) Frames_Skip:{frame_count_skip}"

        # Posiciona o texto - coordenadas inteiras
            cv2.putText(frame, msg_progresso,
                   (int(margin + 10), int(y_pos + bar_height - 7)),
                   cv2.FONT_HERSHEY_SIMPLEX,
                   0.6,
                   (0, 0, 0),
                   2)

        except Exception as e:
            logger.log(f"Erro ao desenhar barra de progresso: {str(e)}")
            return frame

        return frame
```

### 5.4 Limitações do OpenCV e MediaPipe
  
Quando usamos MediaPipe e OpenCV para análise de vídeos pré-gravados, os falsos-positivos, podem ocorrer.

Limitações do MediaPipe Face Mesh/Emotions:

- O MediaPipe pode ter dificuldade em classificar expressões faciais em frames específicos.

- A classificação é feita frame a frame, sem contexto temporal
  Alguns ângulos ou expressões podem confundir o modelo padrão

- Ajustar parâmetros do MediaPipe (min_detection_confidence, min_tracking_confidence)

- Implementar filtro temporal para evitar mudanças bruscas de classificação
Verificar se há frames com baixa qualidade/blur

- Considerar usar outro modelo específico para emoções se a precisão for crítica

Para debug:

- Salvar frames onde há discrepância
- Logar as classificações ao longo do tempo
= Visualizar os landmarks faciais detectados
- O MediaPipe é ótimo para muitos casos, mas pode ter essas inconsistências ocasionais na classificação de emoções.

Parâmetros que afetam a detecção e falsos positivos:

- MEDIAPIPE FACE MESH:
  - static_image_mode: False para vídeo, True para fotos
  - max_num_faces: quantidade máxima de rostos (default 1)
  - min_detection_confidence: 0.5 a 0.9 (maior = mais preciso, menos falsos positivos)
  - min_tracking_confidence: 0.5 a 0.9 (maior = tracking mais estável)

- QUALIDADE DE IMAGEM:
  - Resolução: ideal entre 640x480 e 1280x720
 - FPS: ideal entre 15 e 30
 - Brilho: ajuste entre -50 a +50
 - Contraste: ajuste entre 0.5 a 1.5
 - Blur: kernel 3x3 ou 5x5 para reduzir ruído

- SUAVIZAÇÃO TEMPORAL:
  - Buffer de frames: 3 a 10 frames
  - Threshold confiança: 0.6 a 0.8
 - Frames consistentes: 2 a 5 frames
 - Janela de tempo: 0.1 a 1 segundo

```python
#  16.3 - Transcrição do audio do video

!pip install -q moviepy
!pip install -q git+https://github.com/openai/whisper.git

import whisper
import moviepy.editor as mp

import os
from google.colab import drive
import warnings
import tensorflow as tf

!pip install mediapipe==0.10.0

import pandas as pd
from datetime import datetime
from IPython.display import clear_output
import mediapipe as mp
import subprocess
import sys
import importlib
import platform
import psutil
from moviepy.editor import VideoFileClip
import whisper

file_name='Unlocking_Facial_Recognition_Diverse_Activities_Analysis'
video_path = f'/content/drive/MyDrive/Grupo51_Fiap_Fase4_Challenge/Video/Unlocking_Facial_Recognition_Diverse_Activities_Analysis.mp4'
audio_output_path = f'/content/drive/MyDrive/Grupo51_Fiap_Fase4_Challenge/Video/Audio'

audio_file_name = f'{file_name}.mp3'

def transcribe_video(video_path, output_path):
    try:
        # Extrair áudio do vídeo
        video = VideoFileClip(video_path)
        audio_path = output_path + f"/{audio_file_name}" # Alterado para salvar como MP3
        video.audio.write_audiofile(audio_path, codec='mp3') # Especifica o codec mp3

        # Carregar o modelo Whisper
        model = whisper.load_model("base") # ou outro modelo, se necessário

        # Transcrever o áudio em inglês
        result_en = model.transcribe(audio_path, language='english')

        # Salvar a transcrição em inglês
        transcription_en_path = output_path + "/transcription_english.txt"
        with open(transcription_en_path, "w") as f:
            f.write(result_en["text"])

        print(f"Transcrição em inglês salva em {transcription_en_path}")

        # Transcrever o áudio em português
        result_pt = model.transcribe(audio_path, language='portuguese')

        # Salvar a transcrição em português
        transcription_pt_path = output_path + "/transcription_portuguese.txt"
        with open(transcription_pt_path, "w") as f:
            f.write(result_pt["text"])

        print(f"Transcrição em português salva em {transcription_pt_path}")

    except Exception as e:
        print(f"Ocorreu um erro: {e}")

# Criar o diretório de saída, se não existir
os.makedirs(audio_output_path, exist_ok=True)

transcribe_video(video_path, audio_output_path)
```
